---

## Part 1 — Why Not Just Use a Python List?

```python
# Suppose we have 5 order amounts from our dataset
order_amounts = [4500, 12000, 800, 33000, 7200]

# Let's try to apply 7.5% VAT to every order
order_amounts * 0.075
```

```python
# To make it work with a list, we need a loop
vat_amounts = []
for amount in order_amounts:
    vat_amounts.append(amount * 0.075)

print(vat_amounts)

# This works — but imagine doing this across 500,000 rows.
# Loops visit one element at a time. NumPy does all of them simultaneously.
```

---
## Part 2 — Creating NumPy Arrays

```python
import numpy as np

# np is the universal alias — every data scientist uses this convention
print("NumPy version:", np.__version__)
```

```python
# 1D array — a single sequence of values
daily_revenue = np.array([45000, 62000, 38000, 71000, 55000, 49000, 83000])

print(daily_revenue)
print("Type:",  type(daily_revenue))
print("Shape:", daily_revenue.shape)   # (7,) — 7 elements, 1 dimension
print("Dtype:", daily_revenue.dtype)   # int64 — all integers
print("Ndim: ", daily_revenue.ndim)    # 1 dimension
```

```python
# 2D array — rows and columns, like a mini spreadsheet
# Weekly revenue by product category (Electronics, Fashion, Food)
weekly_revenue = np.array([
    [85000, 42000, 12000],   # Week 1
    [91000, 38000, 15000],   # Week 2
    [78000, 55000, 9000],    # Week 3
    [103000, 61000, 18000],  # Week 4
])

print(weekly_revenue)
print("Shape:", weekly_revenue.shape)  # (4, 3) — 4 rows, 3 columns
print("Ndim: ", weekly_revenue.ndim)   # 2 dimensions
```

---
## Part 3 — Loading the Real Dataset

```python
import csv

# Load our orders dataset manually (we'll use pd.read_csv() in Topic 2)
orders = []
with open("../../../data/orders.csv", encoding="utf-8-sig") as f:
    reader = csv.DictReader(f)
    for row in reader:
        orders.append(row)

print(f"Loaded {len(orders)} orders")
print("Columns:", list(orders[0].keys()))
```

```python
# Extract total_amount as a NumPy array
# Some rows have missing total_amount — skip those for now (Topic 3 will handle them)
total_amounts = np.array([
    float(row["total_amount"])
    for row in orders
    if row["total_amount"] != ""
])

print("Array shape:", total_amounts.shape)
print("First 10 values:", total_amounts[:10])
```

---
## Part 4 — Element-Wise Operations

```python
# Apply 7.5% VAT to every order — no loop needed
vat = total_amounts * 0.075
total_with_vat = total_amounts + vat

print("First 5 original amounts: ", total_amounts[:5])
print("First 5 VAT amounts:      ", vat[:5].round(2))
print("First 5 totals with VAT:  ", total_with_vat[:5].round(2))
```

```python
# Convert Naira to USD (1 USD = 1,600 NGN)
usd_amounts = total_amounts / 1600

print("First 5 amounts in USD:", usd_amounts[:5].round(2))
```

---
## Part 5 — Indexing and Slicing

```python
# Single elements
print("First order: ", total_amounts[0])
print("Last order:  ", total_amounts[-1])
print("5th order:   ", total_amounts[4])
```

```python
# Slices
print("First 7 orders:  ", total_amounts[:7])
print("Orders 10 to 15: ", total_amounts[10:16])
print("Last 5 orders:   ", total_amounts[-5:])
```

```python
# 2D indexing on the weekly_revenue grid from Part 2
print("Week 2, Fashion revenue:      ", weekly_revenue[1, 1])   # row 1, col 1
print("All of Week 3:                ", weekly_revenue[2, :])   # entire row 2
print("All Electronics (column 0):   ", weekly_revenue[:, 0])   # entire column 0
print("Weeks 2 and 3, Fashion & Food:", weekly_revenue[1:3, 1:3])
```

---
## Part 6 — Statistical Functions

```python
print("Total revenue:      NGN", f"{np.sum(total_amounts):,.0f}")
print("Average order:      NGN", f"{np.mean(total_amounts):,.0f}")
print("Median order:       NGN", f"{np.median(total_amounts):,.0f}")
print("Smallest order:     NGN", f"{np.min(total_amounts):,.0f}")
print("Largest order:      NGN", f"{np.max(total_amounts):,.0f}")
print("Std deviation:      NGN", f"{np.std(total_amounts):,.0f}")
print("75th percentile:    NGN", f"{np.percentile(total_amounts, 75):,.0f}")
```

```python
# Why is the mean much higher than the median?
# Let's find those extreme orders
mean_val   = np.mean(total_amounts)
median_val = np.median(total_amounts)

print(f"Mean:   NGN {mean_val:,.0f}")
print(f"Median: NGN {median_val:,.0f}")
print(f"Difference: NGN {mean_val - median_val:,.0f}")
print()

# Orders above NGN 200,000 — the outliers pulling the mean up
outliers = total_amounts[total_amounts > 200000]
print("Extreme orders above NGN 200,000:", outliers)
```

---
## Part 7 — The `axis` Parameter

```python
# Use the weekly_revenue grid: shape (4 weeks, 3 categories)
# Categories: Electronics, Fashion, Food & Drinks
print("Grid shape:", weekly_revenue.shape)
print()

# Total per CATEGORY — sum down the rows (axis=0)
category_totals = np.sum(weekly_revenue, axis=0)
print("Total per category (axis=0):")
print(f"  Electronics: NGN {category_totals[0]:,}")
print(f"  Fashion:     NGN {category_totals[1]:,}")
print(f"  Food:        NGN {category_totals[2]:,}")
print()

# Total per WEEK — sum across the columns (axis=1)
weekly_totals = np.sum(weekly_revenue, axis=1)
print("Total per week (axis=1):")
for i, total in enumerate(weekly_totals, 1):
    print(f"  Week {i}: NGN {total:,}")
```

```python
# Which week had the best performance per category?
best_week_per_category = np.argmax(weekly_revenue, axis=0) + 1  # +1 to make it Week 1-based
print("Best week for Electronics: Week", best_week_per_category[0])
print("Best week for Fashion:     Week", best_week_per_category[1])
print("Best week for Food:        Week", best_week_per_category[2])
```

---
## Part 8 — Boolean Indexing (Filtering)

```python
# Step 1: a condition produces a True/False array
high_value_mask = total_amounts > 20000

print("Boolean mask (first 10):", high_value_mask[:10])
print("Number of high-value orders:", np.sum(high_value_mask))  # True counts as 1
```

```python
# Step 2: use the mask to filter
high_value_orders = total_amounts[total_amounts > 20000]

print("High-value orders (>NGN 20,000):")
print(high_value_orders)
print()
print(f"Count:   {len(high_value_orders)} orders")
print(f"Average: NGN {np.mean(high_value_orders):,.0f}")
```

```python
# Combine conditions with & (and) and | (or)

# Orders between NGN 5,000 and NGN 30,000 (the 'mid-range' band)
mid_range = total_amounts[(total_amounts >= 5000) & (total_amounts <= 30000)]
print(f"Mid-range orders (NGN 5k–30k): {len(mid_range)} orders")
print(f"Average mid-range order: NGN {np.mean(mid_range):,.0f}")
```

---
## Part 9 — Putting It All Together

**Business question:** *"What percentage of our orders are high-value (above NGN 20,000), and how much of total revenue do they represent?"*

```python
all_orders     = total_amounts
high_value     = total_amounts[total_amounts > 20000]
standard_value = total_amounts[total_amounts <= 20000]

total_rev      = np.sum(all_orders)
high_rev       = np.sum(high_value)
standard_rev   = np.sum(standard_value)

print("===== Order Value Analysis =====")
print(f"Total orders:          {len(all_orders)}")
print(f"High-value orders:     {len(high_value)} ({len(high_value)/len(all_orders)*100:.1f}%)")
print(f"Standard orders:       {len(standard_value)} ({len(standard_value)/len(all_orders)*100:.1f}%)")
print()
print(f"Total revenue:         NGN {total_rev:,.0f}")
print(f"High-value revenue:    NGN {high_rev:,.0f} ({high_rev/total_rev*100:.1f}% of total)")
print(f"Standard revenue:      NGN {standard_rev:,.0f} ({standard_rev/total_rev*100:.1f}% of total)")
```

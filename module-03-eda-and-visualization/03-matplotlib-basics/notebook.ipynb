## Live Demo Notebook

**Scenario:** You have two Nigerian datasets — state-level statistics and professional salaries.
Your task: build four core chart types, one at a time, adding proper formatting to each.

---

**My EDA questions:**
1. How do the top states compare by population and GDP?
2. Is there a relationship between a state's population and its GDP?
3. What does the salary distribution look like?

---

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load both datasets
states   = pd.read_csv('nigerian_states.csv')
salaries = pd.read_csv('../../../data/nigerian_salaries.csv')

print('States shape:   ', states.shape)
print('Salaries shape: ', salaries.shape)
states.head(3)
```

---
## Part 1 — The Figure-Axes Structure

Before building any chart, understand the two objects you will always work with.

```python
# An empty figure — just to see the structure
fig, ax = plt.subplots(figsize=(8, 4))

ax.set_title('This is the Axes title (ax.set_title)')
ax.set_xlabel('This is the X-axis label')
ax.set_ylabel('This is the Y-axis label')

plt.tight_layout()
plt.show()

# fig = the outer canvas
# ax  = the plot area — title, labels, and data all go on ax
```

---
## Part 2 — Line Chart

First: a bare chart with no formatting. Then: the same chart properly formatted.

```python
# Top 8 states by population
top8 = states.nlargest(8, 'population_millions').sort_values('population_millions', ascending=False)

# BARE chart — no formatting
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(top8['state'], top8['population_millions'])
plt.show()

# What's missing? Title, axis labels, units, tick rotation
```

```python
# FORMATTED chart — adding everything step by step
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(top8['state'], top8['population_millions'],
        marker='o', color='steelblue', linewidth=2.5, markersize=8)

ax.set_title('Population of Top 8 Nigerian States', fontsize=14, fontweight='bold')
ax.set_xlabel('State', fontsize=12)
ax.set_ylabel('Population (millions)', fontsize=12)
ax.tick_params(axis='x', rotation=30)
ax.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

---
## Part 3 — Bar Chart

Q1: How do the top 10 states compare by GDP?

```python
# Vertical bar — try it first
top10_gdp = states.nlargest(10, 'gdp_billion_naira').sort_values('gdp_billion_naira', ascending=False)

fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(top10_gdp['state'], top10_gdp['gdp_billion_naira'], color='coral', edgecolor='white')
ax.set_title('Top 10 States by GDP (₦ Billion)')
ax.set_xlabel('State')
ax.set_ylabel('GDP (₦ Billion)')
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

# Observation: labels still overlap a bit — let's try horizontal
```

```python
# Horizontal bar — better for long category names
top10_sorted = states.nlargest(10, 'gdp_billion_naira').sort_values('gdp_billion_naira')

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(top10_sorted['state'], top10_sorted['gdp_billion_naira'],
               color='coral', edgecolor='white')

# Add value labels at the end of each bar
for bar in bars:
    width = bar.get_width()
    ax.text(width + 300, bar.get_y() + bar.get_height() / 2,
            f'₦{width:,.0f}B', va='center', fontsize=9)

ax.set_title('Top 10 Nigerian States by GDP (₦ Billion)', fontsize=14, fontweight='bold')
ax.set_xlabel('GDP (₦ Billion)', fontsize=12)
ax.set_ylabel('State', fontsize=12)

plt.tight_layout()
plt.show()
```

---
## Part 4 — Scatter Plot

Q2: Is there a relationship between a state's population and its GDP?

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax.scatter(states['population_millions'], states['gdp_billion_naira'],
           color='seagreen', alpha=0.75, s=90, edgecolors='white', linewidths=0.5)

# Annotate the two highest GDP states
for _, row in states.nlargest(2, 'gdp_billion_naira').iterrows():
    ax.annotate(row['state'],
                xy=(row['population_millions'], row['gdp_billion_naira']),
                xytext=(10, 5), textcoords='offset points',
                fontsize=10, fontweight='bold', color='darkgreen')

ax.set_title('Population vs GDP — All 37 Nigerian States', fontsize=14, fontweight='bold')
ax.set_xlabel('Population (millions)', fontsize=12)
ax.set_ylabel('GDP (₦ Billion)', fontsize=12)
ax.grid(linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()

# What does this chart tell us?
corr = states['population_millions'].corr(states['gdp_billion_naira'])
print(f'Correlation (population vs GDP): {corr:.3f}')
```

---
## Part 5 — Histogram

Q3: What does the Nigerian professional salary distribution look like?

```python
# Try different bin counts to see the effect
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))

for ax, bins, label in zip([ax1, ax2, ax3], [5, 20, 80], ['bins=5 (too few)', 'bins=20 (good)', 'bins=80 (too many)']):
    ax.hist(salaries['salary_monthly'], bins=bins, color='mediumpurple', edgecolor='white', alpha=0.85)
    ax.set_title(label)
    ax.set_xlabel('Salary (₦)')
    ax.set_ylabel('Count')

plt.suptitle('Effect of bin count on histogram shape', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Final histogram with mean and median lines
fig, ax = plt.subplots(figsize=(11, 5))

ax.hist(salaries['salary_monthly'], bins=30, color='mediumpurple',
        edgecolor='white', alpha=0.85)

mean_s   = salaries['salary_monthly'].mean()
median_s = salaries['salary_monthly'].median()

ax.axvline(mean_s,   color='red',    linestyle='--', linewidth=2,
           label=f'Mean: ₦{mean_s:,.0f}')
ax.axvline(median_s, color='orange', linestyle='--', linewidth=2,
           label=f'Median: ₦{median_s:,.0f}')

ax.set_title('Distribution of Nigerian Professional Monthly Salaries', fontsize=14, fontweight='bold')
ax.set_xlabel('Monthly Salary (₦)', fontsize=12)
ax.set_ylabel('Number of Professionals', fontsize=12)
ax.legend(fontsize=11)
ax.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()

print(f'Skewness: {salaries["salary_monthly"].skew():.2f} — right-skewed')
print(f'The mean is ₦{mean_s - median_s:,.0f} higher than the median.')
print('This is the right tail pulling the mean upward.')
```

---
## Part 6 — Subplots: Compare Population and GDP Side by Side

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left — top 6 by population
top6_pop = states.nlargest(6, 'population_millions').sort_values('population_millions')
ax1.barh(top6_pop['state'], top6_pop['population_millions'], color='steelblue')
ax1.set_title('Top 6 by Population', fontsize=12, fontweight='bold')
ax1.set_xlabel('Population (millions)')

# Right — top 6 by GDP
top6_gdp = states.nlargest(6, 'gdp_billion_naira').sort_values('gdp_billion_naira')
ax2.barh(top6_gdp['state'], top6_gdp['gdp_billion_naira'], color='coral')
ax2.set_title('Top 6 by GDP', fontsize=12, fontweight='bold')
ax2.set_xlabel('GDP (₦ Billion)')

fig.suptitle('Nigerian States — Are the Most Populous Also the Richest?', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print('Top 6 by population:', list(top6_pop['state']))
print('Top 6 by GDP:       ', list(top6_gdp['state']))
print()
print('Notice: the lists are not identical. Large population does not guarantee high GDP.')
```

---
## Part 7 — Saving a Chart

```python
fig, ax = plt.subplots(figsize=(10, 5))

top10 = states.nlargest(10, 'gdp_billion_naira').sort_values('gdp_billion_naira')
ax.barh(top10['state'], top10['gdp_billion_naira'], color='coral', edgecolor='white')
ax.set_title('Top 10 Nigerian States by GDP (₦ Billion)', fontsize=14)
ax.set_xlabel('GDP (₦ Billion)')

plt.tight_layout()

# SAVE before SHOW — show() clears the figure from memory
fig.savefig('top10_gdp_states.png', dpi=150, bbox_inches='tight')
print('Chart saved as top10_gdp_states.png')

plt.show()
```

## 18.1 Working with JSON

**JSON (JavaScript Object Notation)** is the dominant data format for exchanging data on the web.

It looks just like Python dictionaries and lists:
- `{}` for objects (Python dict)
- `[]` for arrays (Python list)
- Strings in **double quotes**
- `true`/`false` (lowercase) instead of `True`/`False`

| Function | Direction | Use |
|----------|-----------|-----|
| `json.dumps(obj)` | Python → JSON string | Serialise to string |
| `json.loads(string)` | JSON string → Python | Parse from string |
| `json.dump(obj, file)` | Python → JSON file | Write to file |
| `json.load(file)` | JSON file → Python | Read from file |


````python
# CODE EXAMPLE 18.1a: Converting between Python and JSON

import json

# Python dict -> JSON string
student = {
    'name': 'Amara',
    'score': 85,
    'passed': True,
    'subjects': ['Math', 'English', 'Science']
}

json_str = json.dumps(student, indent=2)   # indent=2 makes it readable
print('JSON string:')
print(json_str)

# JSON string -> Python dict
parsed = json.loads(json_str)
print('\nParsed back to Python:')
print(type(parsed))           # <class 'dict'>
print(parsed['name'])         # Amara
print(parsed['subjects'])     # ['Math', 'English', 'Science']
````

````python
# CODE EXAMPLE 18.1b: Reading and writing JSON files

import json

# Data to save
class_data = [
    {'name': 'Amara',  'score': 85},
    {'name': 'Kofi',   'score': 72},
    {'name': 'Ama',    'score': 91},
    {'name': 'Yaw',    'score': 65},
]

# Write list of dicts to JSON file
with open('class_data.json', 'w') as f:
    json.dump(class_data, f, indent=2)
print('Saved to class_data.json')

# Read it back
with open('class_data.json', 'r') as f:
    loaded = json.load(f)

print(f'Loaded {len(loaded)} students')
for student in loaded:
    print(f"  {student['name']:10s}: {student['score']}")
````

## 18.2 Working with Web APIs

An **API (Application Programming Interface)** allows programs to communicate with web services.  
The `requests` library makes HTTP requests in one line.

**Install:** `pip install requests`

**Key steps:**
1. Call `requests.get(url)` to make a GET request
2. Check `response.status_code` — **200** = success
3. Call `response.json()` to parse the JSON response into a Python dict

**Common status codes:**
- `200 OK` — request succeeded
- `404 Not Found` — resource does not exist
- `401 Unauthorized` — missing or invalid API key
- `500 Internal Server Error` — server-side problem


````python
# CODE EXAMPLE 18.2a: Making a GET request to a public API
# Uses a free public API that requires no API key

import requests

# JSONPlaceholder is a free fake REST API for testing
url = 'https://jsonplaceholder.typicode.com/users/1'

try:
    response = requests.get(url, timeout=10)   # timeout=10s to avoid hanging
    response.raise_for_status()                # raises exception if status >= 400

    user = response.json()                     # parse JSON response
    print(f"Name   : {user['name']}")
    print(f"Email  : {user['email']}")
    print(f"City   : {user['address']['city']}")
    print(f"Company: {user['company']['name']}")

except requests.exceptions.ConnectionError:
    print('No internet connection')
except requests.exceptions.HTTPError as e:
    print(f'HTTP error: {e}')
except requests.exceptions.Timeout:
    print('Request timed out')
````

````python
# CODE EXAMPLE 18.2b: Fetching and filtering a list from an API

import requests

url = 'https://jsonplaceholder.typicode.com/posts'

try:
    response = requests.get(url, timeout=10)
    response.raise_for_status()

    posts = response.json()   # returns a list of 100 posts
    print(f'Total posts fetched: {len(posts)}')

    # Filter: posts by user 1 only
    user1_posts = [p for p in posts if p['userId'] == 1]
    print(f'Posts by user 1   : {len(user1_posts)}')

    # Print first 3 titles
    for post in user1_posts[:3]:
        print(f"  - {post['title']}")

except requests.exceptions.RequestException as e:
    print(f'Request failed: {e}')
````

## 18.3 File Automation with pathlib

`pathlib.Path` (Python 3.4+) provides an object-oriented way to work with file paths.

| Operation | Code |
|-----------|------|
| Current directory | `Path('.')` or `Path.cwd()` |
| Join paths | `Path('folder') / 'subfolder' / 'file.txt'` |
| List files | `Path('.').glob('*.txt')` |
| Check exists | `p.exists()` |
| Create directory | `p.mkdir(parents=True, exist_ok=True)` |
| Read text | `p.read_text()` |
| Write text | `p.write_text('content')` |
| Rename | `p.rename(new_path)` |
| Copy | `shutil.copy(src, dst)` |
| Delete file | `p.unlink()` |


````python
# CODE EXAMPLE 18.3a: Exploring the file system with pathlib

from pathlib import Path

# Current working directory
cwd = Path.cwd()
print('Current directory:', cwd)

# List all Python files in current directory
py_files = list(cwd.glob('*.py'))
print(f'\nPython files found: {len(py_files)}')
for f in py_files[:5]:     # show first 5
    print(f'  {f.name}')

# Path properties
sample = Path('reports/monthly/jan_2024.csv')
print(f'\nPath   : {sample}')
print(f'Name   : {sample.name}')          # jan_2024.csv
print(f'Stem   : {sample.stem}')          # jan_2024
print(f'Suffix : {sample.suffix}')        # .csv
print(f'Parent : {sample.parent}')        # reports/monthly
print(f'Exists : {sample.exists()}')      # False (not created yet)
````

````python
# CODE EXAMPLE 18.3b: Creating directories, writing and reading files

from pathlib import Path
import shutil

# Create a directory structure
output_dir = Path('output') / 'topic18_demo'
output_dir.mkdir(parents=True, exist_ok=True)
print(f'Created: {output_dir}')

# Write a text file
report_file = output_dir / 'report.txt'
report_file.write_text('Topic 18 Report\nCreated with pathlib')
print(f'Written: {report_file}')

# Read it back
content = report_file.read_text()
print(f'Content:\n{content}')

# Copy the file to a backup
backup = output_dir / 'report_backup.txt'
shutil.copy(report_file, backup)
print(f'Backup : {backup} | exists: {backup.exists()}')

# List everything in output_dir
print('\nDirectory contents:')
for item in output_dir.iterdir():
    print(f'  {item.name}')
````

## 18.4 Python Best Practices

Writing code that **works** is the first step. Writing code that is **readable and maintainable** is the professional standard.

### Key Principles

**DRY (Don't Repeat Yourself):** If you copy-paste code, extract it into a function.

**PEP 8:** Python's style guide — 4-space indentation, lowercase_with_underscores for variables, UPPER_CASE for constants.

**Type hints:** `def add(a: int, b: int) -> int:` — document expected types.

**Docstrings:** `"""One-line description."""` — explain what a function does.

**Context managers:** Use `with` for files and connections — guarantees cleanup.

**Guard clause:** `if __name__ == '__main__':` — prevents code running on import.


````python
# CODE EXAMPLE 18.4a: Applying DRY and type hints

# BAD: repeated logic
score1 = 85
if score1 >= 80: grade1 = 'A'
elif score1 >= 70: grade1 = 'B'
else: grade1 = 'C'

score2 = 72
if score2 >= 80: grade2 = 'A'
elif score2 >= 70: grade2 = 'B'
else: grade2 = 'C'

# GOOD: extract to a function (DRY)
def get_grade(score: float) -> str:
    """Return letter grade for score 0-100."""
    if score >= 80: return 'A'
    if score >= 70: return 'B'
    if score >= 60: return 'C'
    if score >= 50: return 'D'
    return 'F'

scores = [85, 72, 91, 54, 67]
for s in scores:
    print(f'{s} -> {get_grade(s)}')
````

````python
# CODE EXAMPLE 18.4b: Context managers and guard clause

import json
from pathlib import Path

# Context manager: file is always closed even if an error occurs
def read_json_safe(filepath: str) -> dict | None:
    """Read a JSON file and return parsed data, or None on error."""
    try:
        with open(filepath, 'r') as f:   # 'with' handles close()
            return json.load(f)
    except FileNotFoundError:
        print(f'File not found: {filepath}')
        return None
    except json.JSONDecodeError:
        print(f'Invalid JSON in: {filepath}')
        return None

def main() -> None:
    """Main entry point."""
    data = read_json_safe('class_data.json')
    if data:
        print(f'Loaded {len(data)} records')

# Guard clause: only runs when this file is executed directly
if __name__ == '__main__':
    main()
````

## 18.5 Building a Command-Line Application

A **command-line application (CLI)** is a program that users run from the terminal.

Python's `argparse` module handles arguments and help text automatically:

```bash
# Example usage from terminal:
python report.py students.json --format summary
```

**Key `argparse` components:**
- `ArgumentParser(description='...')` — creates the parser with a description
- `parser.add_argument('name')` — positional argument (required)
- `parser.add_argument('--flag')` — optional argument
- `parser.parse_args()` — reads sys.argv and returns a Namespace object

**`if __name__ == '__main__':`** — ensures the script only runs when executed directly, not when imported.


````python
# CODE EXAMPLE 18.5a: Building a simple CLI with argparse
# Note: argparse reads sys.argv, so in Jupyter we simulate args

import argparse
import sys

def build_parser() -> argparse.ArgumentParser:
    """Build and return the argument parser."""
    parser = argparse.ArgumentParser(
        description='Student Report Tool'
    )
    parser.add_argument('name', help='Student name')
    parser.add_argument('score', type=float, help='Exam score (0-100)')
    parser.add_argument('--verbose', action='store_true',
                        help='Show extra details')
    return parser

def get_grade(score: float) -> str:
    """Return letter grade."""
    if score >= 80: return 'A'
    if score >= 70: return 'B'
    if score >= 60: return 'C'
    if score >= 50: return 'D'
    return 'F'

# Simulate running from terminal: python report.py Amara 85 --verbose
sys.argv = ['report.py', 'Amara', '85', '--verbose']

parser = build_parser()
args = parser.parse_args()

grade = get_grade(args.score)
print(f'Student: {args.name}')
print(f'Score  : {args.score}')
print(f'Grade  : {grade}')
if args.verbose:
    print(f'Status : {"Pass" if args.score >= 50 else "Fail"}')
````

````python
# CODE EXAMPLE 18.5b: Capstone — JSON report generator CLI

import json
from pathlib import Path

def load_students(path: str) -> list:
    """Load student list from a JSON file."""
    with open(path, 'r') as f:
        return json.load(f)

def enrich_students(students: list) -> list:
    """Add grade and status fields to each student dict."""
    grade_map = [(80, 'A'), (70, 'B'), (60, 'C'), (50, 'D')]
    for s in students:
        s['grade'] = next(
            (g for threshold, g in grade_map if s['score'] >= threshold),
            'F'
        )
        s['status'] = 'Pass' if s['score'] >= 50 else 'Fail'
    return students

def save_report(students: list, output_path: str) -> None:
    """Save enriched student list to JSON."""
    Path(output_path).parent.mkdir(parents=True, exist_ok=True)
    with open(output_path, 'w') as f:
        json.dump(students, f, indent=2)
    print(f'Report saved → {output_path}')

def print_summary(students: list) -> None:
    """Print a formatted console summary."""
    print('\n=== Class Report ===')
    print(f'{"Name":<12} {"Score":>6} {"Grade":>6} {"Status":>8}')
    print('-' * 38)
    for s in sorted(students, key=lambda x: x['score'], reverse=True):
        print(f"{s['name']:<12} {s['score']:>6.1f} {s['grade']:>6} {s['status']:>8}")
    scores = [s['score'] for s in students]
    print('-' * 38)
    print(f'Average: {sum(scores)/len(scores):.1f} | Pass: {sum(1 for s in students if s["status"]=="Pass")}/{len(students)}')

# Run the pipeline
students = load_students('class_data.json')   # created in 18.1b
students = enrich_students(students)
save_report(students, 'output/class_report.json')
print_summary(students)
````

## 4.1 Arithmetic Operators

Python has 7 arithmetic operators:

- `+` addition, `-` subtraction, `*` multiplication
- `/` **true division** — always returns a float (e.g., `6/2` → `3.0`)
- `//` **floor division** — divides and drops the decimal (e.g., `7//2` → `3`)
- `%` **modulo** — returns the remainder (e.g., `7%2` → `1`)
- `**` **exponentiation** — raises to a power (e.g., `2**8` → `256`)

The `%` (modulo) is especially useful for checking if a number is even: `if n % 2 == 0`


````python
# CODE EXAMPLE 4.1a: All arithmetic operators

a = 17
b = 5

print("Addition:       ", a + b)   # 22
print("Subtraction:    ", a - b)   # 12
print("Multiplication: ", a * b)   # 85
print("True division:  ", a / b)   # 3.4  (always float)
print("Floor division: ", a // b)  # 3    (drops decimal)
print("Modulo:         ", a % b)   # 2    (remainder)
print("Exponentiation: ", a ** 2)  # 289
````

````python
# CODE EXAMPLE 4.1b: Practical uses

# Modulo: even/odd check
number = 24
if number % 2 == 0:
    print(number, "is even")
else:
    print(number, "is odd")

# Floor division: calculate how many complete weeks in 50 days
days = 50
complete_weeks = days // 7
remaining_days = days % 7
print(f"{days} days = {complete_weeks} complete weeks and {remaining_days} remaining days")

# Exponentiation: compound interest base
principal = 1000
rate = 1.05   # 5% interest
amount_after_3_years = principal * (rate ** 3)
print(f"After 3 years at 5% interest: {amount_after_3_years:.2f}")
````

## 4.2 Assignment Operators

Assignment operators combine an operation and assignment in one step.

- `x += 5` is exactly the same as `x = x + 5`
- `x -= 3` is the same as `x = x - 3`
- `x *= 2` is the same as `x = x * 2`
- `x //= 4` is the same as `x = x // 4`

**Common mistake:** Writing `=+` instead of `+=` — this just assigns a positive number (e.g., `x =+ 5` sets x to positive 5)


````python
# CODE EXAMPLE 4.2a: Assignment operators in action

# Simulating a bank account balance
balance = 500.00
print("Starting balance:", balance)

balance += 200.00   # deposit
print("After deposit:   ", balance)   # 700.00

balance -= 150.00   # withdrawal
print("After withdrawal:", balance)   # 550.00

# Apply 2% monthly interest
balance *= 1.02
print("After interest:  ", round(balance, 2))  # 561.0
````

## 4.3 Comparison Operators

Comparison operators always return `True` or `False`.

- `==` equal to, `!=` not equal to
- `<` less than, `>` greater than
- `<=` less than or equal to, `>=` greater than or equal to

**Remember:** `=` assigns a value; `==` compares two values.

**String comparisons are case-sensitive:** `"hello" == "Hello"` is `False`.


````python
# CODE EXAMPLE 4.3a: Comparison operators

temperature = 38

print(temperature > 30)    # True  — above 30?
print(temperature == 37)   # False — exactly 37?
print(temperature != 36)   # True  — not 36?
print(temperature <= 40)   # True  — at most 40?

# Comparison result is always bool
is_fever = temperature >= 38
print("Is fever?", is_fever)           # True
print("Type:", type(is_fever))         # <class 'bool'>

# Case-sensitive string comparison
print("python" == "Python")            # False
````

## 4.4 Logical Operators

Logical operators combine boolean expressions.

- `and` — True only when **both** sides are True
- `or` — True when **at least one** side is True
- `not` — **inverts** the boolean

**Critical mistake to avoid:**
```python
# WRONG — '6' is always truthy, so this is always True
if x == 5 or 6:

# CORRECT — each side has a full comparison
if x == 5 or x == 6:
```


````python
# CODE EXAMPLE 4.4a: Logical operators

age = 22
has_id = True
is_banned = False

# 'and' — both must be true
can_enter = age >= 18 and has_id and not is_banned
print("Can enter venue:", can_enter)   # True

# 'or' — at least one must be true
needs_attention = age < 18 or is_banned
print("Needs attention:", needs_attention)  # False

# Correct way to check multiple values
grade = "B"
is_pass_grade = grade == "A" or grade == "B" or grade == "C"
print("Is pass grade:", is_pass_grade)  # True
````

## 4.5 Operator Precedence

Python evaluates operators in this order (highest to lowest):

1. `()` Parentheses — always first
2. `**` Exponentiation
3. `* / // %` Multiplication and division
4. `+ -` Addition and subtraction
5. `== != < > <= >=` Comparisons
6. `not` then `and` then `or` — logical (lowest)

**Best practice:** When in doubt, add parentheses. They also make code more readable.


````python
# CODE EXAMPLE 4.5a: Operator precedence

# Without parentheses — follows rules
result1 = 2 + 3 * 4      # 14  (* before +)
result2 = 2 ** 3 + 1     # 9   (** before +)
result3 = 10 - 2 * 3 + 1 # 5   (* first, then left to right)

print(result1, result2, result3)

# With parentheses — you control the order
result4 = (2 + 3) * 4    # 20
result5 = (2 ** 3) + 1   # 9  (same result, but clearer intent)
print(result4, result5)

# Complex expression — parentheses add clarity
score = 75
attendance = 0.9
# Is this student eligible? Score over 70 AND attendance over 80%
eligible = (score >= 70) and (attendance >= 0.8)
print("Eligible:", eligible)
````

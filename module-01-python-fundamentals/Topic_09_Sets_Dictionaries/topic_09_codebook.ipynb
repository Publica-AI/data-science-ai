## 9.1 What Sets Are

A **set** is an unordered collection of **unique** elements.

Key properties:
- No duplicates — adding an existing item has no effect
- No guaranteed order — cannot access by index
- Very fast membership testing with `in`
- Create with `{1, 2, 3}` OR `set([1, 2, 3])`
- **Empty set MUST use `set()`** — `{}` creates an empty dict


```python
# CODE EXAMPLE 9.1a: Sets basics

# Duplicates are automatically removed
numbers = {1, 2, 3, 2, 1, 3, 4}
print(numbers)              # {1, 2, 3, 4} — duplicates gone

# Remove duplicates from a list
raw_list = ["python", "java", "python", "c++", "java"]
unique_languages = set(raw_list)
print(unique_languages)     # {'python', 'java', 'c++'}

# Membership testing (fast)
print("python" in unique_languages)    # True
print("ruby" in unique_languages)      # False

# Empty set
empty_set = set()           # NOT {}
print(type(empty_set))      # <class 'set'>
```

```python
# CODE EXAMPLE 9.1b: Set methods

my_set = {1, 2, 3}
my_set.add(4)               # add a single item
my_set.add(2)               # already there — no change
print(my_set)               # {1, 2, 3, 4}

my_set.discard(3)           # remove if present (no error if absent)
my_set.remove(1)            # remove (raises KeyError if absent)
print(my_set)               # {2, 4}
```

## 9.2 Set Operations

Sets support mathematical operations — think Venn diagrams:

- **Union** (`|` or `.union()`) — ALL elements from both sets
- **Intersection** (`&` or `.intersection()`) — elements in BOTH sets
- **Difference** (`-` or `.difference()`) — elements in first but NOT second
- **Symmetric difference** (`^`) — elements in either but NOT both


```python
# CODE EXAMPLE 9.2a: Set operations

python_devs = {"Amara", "Kofi", "Ama", "Yaw"}
java_devs = {"Kofi", "Ama", "Abena", "Kojo"}

# Union — all developers
all_devs = python_devs | java_devs
print("All:", all_devs)

# Intersection — know both
both = python_devs & java_devs
print("Both Python and Java:", both)     # {'Kofi', 'Ama'}

# Difference — Python only
python_only = python_devs - java_devs
print("Python only:", python_only)       # {'Amara', 'Yaw'}

# Symmetric difference — not in both
exclusive = python_devs ^ java_devs
print("Not shared:", exclusive)
```

## 9.3 Creating and Accessing Dictionaries

A **dictionary** stores data as key-value pairs.

- Keys must be unique and immutable (strings, numbers, tuples)
- Access with `dict[key]` — raises `KeyError` if key missing
- Safer: `dict.get(key)` — returns `None` or a default if missing
- Check existence: `key in dict`


```python
# CODE EXAMPLE 9.3a: Creating and accessing dictionaries

student = {
    "name": "Amara Mensah",
    "id": 10245,
    "gpa": 3.8,
    "enrolled": True
}

# Access by key
print(student["name"])           # Amara Mensah
print(student["gpa"])            # 3.8

# Safe access with .get()
print(student.get("phone"))      # None (no crash)
print(student.get("phone", "N/A")) # N/A (custom default)

# Check key existence
print("name" in student)         # True
print("email" in student)        # False

# All keys and values
print(list(student.keys()))      # ['name', 'id', 'gpa', 'enrolled']
print(list(student.values()))
```

## 9.4 Modifying Dictionaries

- **Add:** `d["new_key"] = value`
- **Update:** `d["existing_key"] = new_value` (same syntax!)
- **Delete:** `del d["key"]` or `d.pop("key")` (pop returns the value)
- **Merge:** `d.update(other_dict)` — adds all from other dict


```python
# CODE EXAMPLE 9.4a: Modifying dictionaries

profile = {"name": "Kofi", "age": 25}

# Add new key
profile["city"] = "Accra"
print(profile)

# Update existing key
profile["age"] = 26
print(profile)

# Delete with del
del profile["city"]
print(profile)

# Delete with pop (returns the removed value)
old_age = profile.pop("age")
print(f"Removed age: {old_age}")
print(profile)

# Update with another dict
extras = {"email": "kofi@example.com", "age": 27}
profile.update(extras)
print(profile)
```

## 9.5 Iterating Over Dictionaries

Three ways to iterate:

1. `for key in d:` — gives keys only
2. `for value in d.values():` — gives values only
3. `for key, value in d.items():` — gives both (most useful)

**Common mistake:** `for item in d:` — `item` is the KEY, not the value!


```python
# CODE EXAMPLE 9.5a: Iterating dictionaries

scores = {"Amara": 85, "Kofi": 72, "Ama": 91, "Yaw": 65}

# Keys only
print("Students:", list(scores.keys()))

# Values only — calculate average
total = sum(scores.values())
avg = total / len(scores)
print(f"Average: {avg:.1f}")

# Both — most common pattern
print("\nResults:")
for name, score in scores.items():
    grade = "PASS" if score >= 70 else "FAIL"
    print(f"  {name}: {score} — {grade}")
```

## Topic 9 Recap

**Set:** `{a, b, c}` — unique, unordered; `set()` for empty; `in` for fast lookup.

**Set ops:** `|` union, `&` intersection, `-` difference, `^` symmetric difference.

**Dict:** `{"key": value}` — key-value pairs; access by key; `.get()` for safe access.

**Dict modify:** `d[key] = val` adds/updates; `del d[key]` or `.pop(key)` removes.

**Dict iterate:** `.keys()`, `.values()`, `.items()` — use `.items()` for key+value.

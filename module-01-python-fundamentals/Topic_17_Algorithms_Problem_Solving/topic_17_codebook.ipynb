## 17.1 What is an Algorithm?

An **algorithm** is a step-by-step procedure for solving a problem.

Every algorithm must be:
- **Correct** — produces the right output for every valid input
- **Finite** — it must eventually terminate
- **Unambiguous** — each step is clearly defined
- **Efficient** — avoids unnecessary work

The same problem can often be solved by many different algorithms.  
Choosing the right one can mean the difference between a program finishing in **1 second** vs **1 day**.


````python
# CODE EXAMPLE 17.1a: A simple algorithm — find the maximum value
# Algorithm properties: clear steps, finite loop, correct output

def find_max(numbers):
    """
    Input  : a list of numbers
    Process: compare each element against the current maximum
    Output : the largest number in the list
    """
    max_val = numbers[0]        # step 1: assume first element is the maximum
    for num in numbers[1:]:     # step 2: iterate through remaining elements
        if num > max_val:       # step 3: is this number larger?
            max_val = num       # step 4: yes — update the maximum
    return max_val              # step 5: return the result

data = [3, 7, 2, 9, 4, 11, 1]
print('List :', data)
print('Max  :', find_max(data))   # expected: 11
````

````python
# CODE EXAMPLE 17.1b: Counting occurrences — another everyday algorithm

def count_occurrences(lst, target):
    """
    Input  : a list and a target value
    Process: check every element for a match
    Output : how many times target appears
    """
    count = 0                   # initialise counter to zero
    for item in lst:            # visit every element
        if item == target:      # found a match?
            count += 1          # increment counter
    return count

fruits = ['mango', 'orange', 'mango', 'apple', 'mango', 'pear']
print('mangoes found:', count_occurrences(fruits, 'mango'))   # 3
print('apples found :', count_occurrences(fruits, 'apple'))   # 1
````

## 17.2 Searching Algorithms

### Linear Search
- Checks **every element** from start to end
- Works on **any list** (sorted or unsorted)
- Simple but slow for large data — **O(n)**

### Binary Search
- **Requires a sorted list**
- Repeatedly cuts the search space in half
- Very fast — **O(log n)**
- With 1,000,000 items: linear may check 1,000,000 items; binary checks at most **20**


````python
# CODE EXAMPLE 17.2a: Linear Search

def linear_search(lst, target):
    """Check every element; return index or -1 if not found."""
    for i, item in enumerate(lst):   # enumerate gives (index, value)
        if item == target:
            return i                 # found — return the position
    return -1                        # not found

names = ['Amara', 'Kofi', 'Ama', 'Yaw', 'Abena', 'Kojo']
print(linear_search(names, 'Yaw'))    # 3
print(linear_search(names, 'Akua'))   # -1 (not in list)
````

````python
# CODE EXAMPLE 17.2b: Binary Search

def binary_search(sorted_lst, target):
    """Divide and conquer search; list MUST be sorted."""
    low  = 0
    high = len(sorted_lst) - 1

    while low <= high:
        mid = (low + high) // 2          # find the middle index

        if sorted_lst[mid] == target:
            return mid                   # found it!
        elif sorted_lst[mid] < target:
            low = mid + 1                # target is in the RIGHT half
        else:
            high = mid - 1              # target is in the LEFT half

    return -1                           # not found

numbers = [2, 5, 8, 12, 16, 23, 38, 45, 67, 90]
print(binary_search(numbers, 23))    # 5  — found at index 5
print(binary_search(numbers, 10))    # -1 — not in list

# Comparison: both find the same answer
print(linear_search(numbers, 45))    # 7
print(binary_search(numbers, 45))    # 7
````

## 17.3 Sorting — Bubble Sort

Sorting means putting elements in order (ascending or descending).

**Bubble Sort** works by:
1. Comparing adjacent pairs
2. Swapping them if they are in the wrong order
3. After each full pass, the **largest unsorted element** has "bubbled" to its final position
4. Repeating until no swaps occur

**Time complexity:** O(n²) — use Python's built-in `sorted()` for real work.  
Understanding bubble sort builds intuition for **all** sorting algorithms.


````python
# CODE EXAMPLE 17.3a: Bubble Sort with step-by-step tracing

def bubble_sort(lst):
    n = len(lst)
    for i in range(n):                  # outer pass (n passes max)
        swapped = False
        for j in range(0, n - i - 1):  # inner compare (shrinks each pass)
            if lst[j] > lst[j + 1]:    # wrong order?
                lst[j], lst[j + 1] = lst[j + 1], lst[j]  # swap
                swapped = True
        if not swapped:                # no swaps = already sorted
            break
    return lst

data = [64, 34, 25, 12, 22, 11, 90]
print('Before:', data)
result = bubble_sort(data.copy())       # .copy() keeps original intact
print('After :', result)
````

## 17.4 Recursion

**Recursion** is when a function calls **itself** to solve a smaller version of the same problem.

Every recursive function needs:
1. **Base case** — the simplest version; does NOT recurse; just returns a value
2. **Recursive case** — calls itself with a smaller input, moving toward the base case

Without a base case, recursion goes on forever → **RecursionError** (stack overflow).


````python
# CODE EXAMPLE 17.4a: Factorial and Fibonacci using recursion

def factorial(n):
    """n! = n × (n-1) × ... × 1"""
    if n == 0:                     # base case: 0! = 1
        return 1
    return n * factorial(n - 1)    # recursive case

def fibonacci(n):
    """fib(n) = fib(n-1) + fib(n-2)"""
    if n <= 1:                     # base cases: fib(0)=0, fib(1)=1
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)   # recursive case

# Factorial
for i in range(6):
    print(f'{i}! = {factorial(i)}')

# First 10 Fibonacci numbers
print('\nFibonacci:', [fibonacci(i) for i in range(10)])
````

````python
# CODE EXAMPLE 17.4b: Recursive sum and countdown

def recursive_sum(lst):
    """Sum all numbers in a list using recursion."""
    if len(lst) == 0:              # base case: empty list sums to 0
        return 0
    return lst[0] + recursive_sum(lst[1:])  # add first + sum of rest

def countdown(n):
    """Count down from n to 0."""
    if n < 0:                      # base case: stop below 0
        print('Liftoff!')
        return
    print(n)
    countdown(n - 1)               # recursive case

print('Sum of [1,2,3,4,5]:', recursive_sum([1, 2, 3, 4, 5]))  # 15
print()
countdown(5)
````

## 17.5 Big O Notation — Measuring Efficiency

**Big O** describes how an algorithm's runtime grows as input size (n) grows.

| Notation | Name | Example |
|---|---|---|
| O(1) | Constant | Accessing a list by index |
| O(log n) | Logarithmic | Binary search |
| O(n) | Linear | Linear search, find_max |
| O(n log n) | Log-linear | Python's built-in sort |
| O(n²) | Quadratic | Bubble sort, nested loops |

**Rule:** O(1) < O(log n) < O(n) < O(n log n) < O(n²)  
Choose a better algorithm **before** writing the code.


````python
# CODE EXAMPLE 17.5a: Identifying Big O classes

# O(1) — constant time: same speed regardless of list size
def get_first(lst):
    return lst[0]

# O(n) — linear: one operation per element
def find_max_linear(lst):
    max_val = lst[0]
    for item in lst:       # one loop = O(n)
        if item > max_val:
            max_val = item
    return max_val

# O(n²) — quadratic: nested loops
def has_duplicate(lst):
    for i in range(len(lst)):           # outer loop: n times
        for j in range(i + 1, len(lst)):# inner loop: up to n times
            if lst[i] == lst[j]:
                return True
    return False

test = [4, 2, 7, 1, 9]
print('First   :', get_first(test))          # O(1)
print('Max     :', find_max_linear(test))     # O(n)
print('Dup?    :', has_duplicate(test))       # O(n²)
````

````python
# CODE EXAMPLE 17.5b: Big O in practice — comparing approaches

import time

# O(n²): check every pair for duplicates
def has_dup_slow(lst):
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if lst[i] == lst[j]:
                return True
    return False

# O(n): use a set to track seen values
def has_dup_fast(lst):
    seen = set()
    for item in lst:
        if item in seen:    # set lookup is O(1)
            return True
        seen.add(item)
    return False

large = list(range(5000))   # no duplicates in 0..4999

start = time.time()
has_dup_slow(large)
slow_t = time.time() - start

start = time.time()
has_dup_fast(large)
fast_t = time.time() - start

print(f'O(n²) approach: {slow_t:.4f}s')
print(f'O(n)  approach: {fast_t:.4f}s')
print(f'Speed-up: {slow_t / fast_t:.0f}x faster')
````

## Topic 17 Recap

**Algorithm** — step-by-step procedure: correct, finite, unambiguous, efficient.

**Linear Search** — O(n), any list; check every element until found.  
**Binary Search** — O(log n), **sorted list only**; halve the search space each step.

**Bubble Sort** — O(n²); swap adjacent elements repeatedly; for real work use `sorted()`.  

**Recursion** — function calls itself; always define the **base case first**; recursive case moves toward base case.

**Big O** — growth rate: O(1) < O(log n) < O(n) < O(n log n) < O(n²).  
Think algorithm first, code second — a better algorithm beats any hardware upgrade.

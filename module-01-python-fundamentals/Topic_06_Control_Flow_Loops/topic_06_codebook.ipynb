## 6.1 The while Loop

The `while` loop repeats a block of code as long as a condition is `True`.

```python
while condition:
    # block runs while condition is True
    # MUST update something to avoid infinite loop
```

**Key rule:** The loop body must eventually make the condition `False`. Forgetting to update the loop variable creates an **infinite loop**.

**Best use case:** When you do not know in advance how many iterations are needed.


````python
# CODE EXAMPLE 6.1a: Basic while loop with counter

count = 1

while count <= 5:
    print(f"Iteration {count}")
    count += 1          # update counter — critical!

print("Loop finished. count =", count)
````

````python
# CODE EXAMPLE 6.1b: Input validation with while

# Keep asking until valid input is given
age = int(input("Enter your age (must be > 0): "))

while age <= 0:
    print("Invalid age. Please enter a positive number.")
    age = int(input("Enter your age again: "))

print(f"Age recorded: {age}")
````

## 6.2 The for Loop

The `for` loop iterates over a sequence, processing each item one at a time.

```python
for item in sequence:
    # runs once for each item
```

- The loop variable (`item`) automatically takes each value from the sequence
- Works with lists, strings, tuples, dictionaries, and `range()`
- Preferred over `while` when iterating over a known collection
- Use `enumerate()` when you need both the index and the item


````python
# CODE EXAMPLE 6.2a: for loop over a list

students = ["Amara", "Kofi", "Ama", "Kweku"]

# Iterate over the list
for student in students:
    print(f"Hello, {student}!")

print()

# enumerate() gives index + value
for index, student in enumerate(students):
    print(f"{index + 1}. {student}")
````

````python
# CODE EXAMPLE 6.2b: for loop over a string

word = "Python"

# Iterate character by character
for char in word:
    print(char, end="-")   # P-y-t-h-o-n-

print()  # new line

# Count vowels
vowels = "aeiouAEIOU"
sentence = "Programming is fun"
count = 0

for char in sentence:
    if char in vowels:
        count += 1

print(f"Vowels in '{sentence}': {count}")
````

## 6.3 The range() Function

- `range(stop)` — 0 to stop-1 (stop is exclusive)
- `range(start, stop)` — start to stop-1
- `range(start, stop, step)` — start to stop-1 with step increment

**Remember:** `range(10)` gives 0-9, NOT 0-10.


````python
# CODE EXAMPLE 6.3a: range() variants

print("range(5):")
for i in range(5):
    print(i, end=" ")    # 0 1 2 3 4

print("\nrange(1, 6):")
for i in range(1, 6):
    print(i, end=" ")    # 1 2 3 4 5

print("\nrange(0, 20, 5):")
for i in range(0, 20, 5):
    print(i, end=" ")    # 0 5 10 15

print("\nCounting down range(5, 0, -1):")
for i in range(5, 0, -1):
    print(i, end=" ")    # 5 4 3 2 1
````

````python
# CODE EXAMPLE 6.3b: Practical range usage

# Calculate sum of 1 to 100
total = 0
for i in range(1, 101):  # 1 to 100 inclusive
    total += i
print("Sum 1–100:", total)  # 5050

# Print 5-times table
number = 5
for i in range(1, 13):   # 1 to 12
    print(f"{number} x {i:2d} = {number * i:3d}")
````

## 6.4 break and continue

- **`break`** — immediately exits the loop; skips remaining iterations and the else clause
- **`continue`** — skips the rest of the current iteration; jumps to the next one

Both are always used inside an `if` statement inside a loop.


````python
# CODE EXAMPLE 6.4a: break and continue

# break: stop the loop early
print("Finding first multiple of 7 above 20:")
for n in range(21, 100):
    if n % 7 == 0:
        print("Found:", n)
        break              # stop as soon as we find it

print()

# continue: skip certain iterations
print("Odd numbers from 1 to 15:")
for n in range(1, 16):
    if n % 2 == 0:
        continue           # skip even numbers
    print(n, end=" ")      # only prints odd numbers
````

````python
# CODE EXAMPLE 6.4b: Combining all loop concepts

# Find the first 3 numbers divisible by both 3 and 5
found = 0
n = 1

while n <= 1000:
    if n % 3 != 0 or n % 5 != 0:
        n += 1
        continue           # skip if not divisible by both
    print(n)               # divisible by both 3 and 5
    found += 1
    n += 1
    if found == 3:
        break              # stop after finding 3
````

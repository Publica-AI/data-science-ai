## 16.2 NumPy Arrays

- `np.array(list)` — create a 1D array
- `np.arange(start, stop, step)` — range as array
- `np.zeros(n)`, `np.ones(n)` — filled arrays
- `np.random.randint(low, high, size)` — random integer array
- Vectorized operations: `arr * 2`, `arr + 5` — applied to every element
- Statistics: `np.mean()`, `np.std()`, `np.min()`, `np.max()`, `np.sum()`


```python
# CODE EXAMPLE 16.2a: Creating and manipulating arrays

import numpy as np

# Create arrays
arr1 = np.array([10, 20, 30, 40, 50])
arr2 = np.arange(0, 50, 10)          # [0 10 20 30 40]
zeros = np.zeros(5)                  # [0. 0. 0. 0. 0.]
rand = np.random.randint(1, 101, 5)  # 5 random ints 1-100

print('arr1:', arr1)
print('arange:', arr2)
print('zeros:', zeros)
print('random:', rand)

# Vectorized operations — no loop needed
print('doubled:', arr1 * 2)
print('plus 5 :', arr1 + 5)
print('squared:', arr1 ** 2)
print('sum two:', arr1 + arr2)
```

```python
# CODE EXAMPLE 16.2b: NumPy statistics

import numpy as np

scores = np.array([72, 85, 91, 65, 78, 88, 70, 95, 62, 83])

print(f'Count  : {len(scores)}')
print(f'Mean   : {np.mean(scores):.2f}')
print(f'Median : {np.median(scores):.2f}')
print(f'Std Dev: {np.std(scores):.2f}')
print(f'Min    : {np.min(scores)}')
print(f'Max    : {np.max(scores)}')

# Boolean indexing
passing = scores[scores >= 70]
print(f'Passing scores: {passing}')
print(f'Pass rate: {len(passing)/len(scores)*100:.1f}%')
```

## 16.3 Pandas DataFrames

- `pd.DataFrame(dict)` — create from dictionary (keys = column names)
- `df['column']` — select one column (returns Series)
- `df[['col1', 'col2']]` — select multiple columns
- `df[df['col'] > value]` — filter rows (boolean indexing)
- `df.describe()` — statistics summary
- `df.head(n)` — first n rows; `df.tail(n)` — last n rows
- `df.sort_values('col')` — sort by column


```python
# CODE EXAMPLE 16.3a: Creating and exploring a DataFrame

import pandas as pd

data = {
    'name':   ['Amara', 'Kofi', 'Ama', 'Yaw', 'Abena'],
    'score':  [85, 72, 91, 65, 88],
    'grade':  ['A', 'B', 'A', 'C', 'A'],
    'passed': [True, True, True, False, True]
}
df = pd.DataFrame(data)

print(df)
print('\nShape:', df.shape)              # (5, 4)
print('Columns:', df.columns.tolist())
print('\nScore stats:')
print(df['score'].describe())
```

```python
# CODE EXAMPLE 16.3b: Filtering, sorting, and aggregating

import pandas as pd

data = {'name': ['Amara','Kofi','Ama','Yaw','Abena'],
        'score': [85, 72, 91, 65, 88],
        'grade': ['A','B','A','C','A']}
df = pd.DataFrame(data)

# Filter
top = df[df['score'] >= 80]
print('Top students:')
print(top)

# Sort
ranked = df.sort_values('score', ascending=False)
print('\nRanked:')
print(ranked[['name', 'score']].to_string(index=False))

# Aggregation
print('\nGrade distribution:')
print(df['grade'].value_counts())
print(f'Class average: {df["score"].mean():.1f}')
```

## 16.4 Reading CSV Files with Pandas

- `pd.read_csv('file.csv')` — load CSV into a DataFrame
- `df.head()` — first 5 rows; `df.info()` — column types and nulls
- `df.describe()` — statistical summary of numeric columns
- `df['col'].value_counts()` — frequency of each unique value
- `df.isnull().sum()` — count missing values per column
- `df.to_csv('output.csv', index=False)` — save DataFrame to CSV


```python
# CODE EXAMPLE 16.4a: Create and read a CSV with Pandas

import pandas as pd

# Create a CSV to work with
data = {'name': ['Amara','Kofi','Ama','Yaw','Abena','Kojo'],
        'score': [85, 72, 91, 65, 88, 77],
        'grade': ['A','B','A','C','A','B']}
pd.DataFrame(data).to_csv('students_data.csv', index=False)

# Read it back
df = pd.read_csv('students_data.csv')
print(df.head())
print('\nInfo:')
df.info()
print('\nDescribe:')
print(df.describe())
```

```python
# CODE EXAMPLE 16.4b: Analysis on a CSV dataset

import pandas as pd

df = pd.read_csv('students_data.csv')

print('Grade distribution:')
print(df['grade'].value_counts())

print(f'\nHighest scorer: {df.loc[df["score"].idxmax(), "name"]} ({df["score"].max()})')
print(f'Class average : {df["score"].mean():.1f}')
print(f'Pass rate     : {(df["score"] >= 50).sum()}/{len(df)}')

# Save filtered results
df[df['grade'] == 'A'].to_csv('a_students.csv', index=False)
print('\nA-grade students saved to a_students.csv')
```

## 16.5 Matplotlib Basics

- `plt.plot(x, y)` — line plot
- `plt.bar(x, y)` / `plt.barh(x, y)` — vertical/horizontal bar chart
- `plt.scatter(x, y)` — scatter plot
- `plt.hist(data, bins=10)` — histogram
- `plt.title()`, `plt.xlabel()`, `plt.ylabel()` — labels
- `plt.legend()`, `plt.grid(True)` — extras
- `plt.show()` — display the plot; `plt.savefig('file.png')` — save


```python
# CODE EXAMPLE 16.5a: Line and bar charts

import matplotlib.pyplot as plt
import numpy as np

# Line plot
weeks = np.arange(1, 7)
scores = [60, 65, 72, 78, 85, 91]

plt.figure(figsize=(8, 4))
plt.plot(weeks, scores, marker='o', color='steelblue', linewidth=2)
plt.title('Score Improvement Over 6 Weeks')
plt.xlabel('Week')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()

# Bar chart
names = ['Amara', 'Kofi', 'Ama', 'Yaw']
values = [85, 72, 91, 65]
plt.figure(figsize=(7, 4))
plt.bar(names, values, color=['steelblue','coral','mediumseagreen','orange'])
plt.title('Student Scores')
plt.ylabel('Score')
plt.ylim(0, 100)
plt.show()
```

```python
# CODE EXAMPLE 16.5b: Histogram and scatter plot

import matplotlib.pyplot as plt
import numpy as np

# Histogram
scores = np.random.randint(40, 101, 50)  # 50 random scores
plt.figure(figsize=(7, 4))
plt.hist(scores, bins=10, color='steelblue', edgecolor='black')
plt.title('Score Distribution (50 Students)')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

# Scatter plot
study_hours = np.random.uniform(1, 8, 20)
exam_scores = study_hours * 10 + np.random.normal(0, 5, 20)
plt.figure(figsize=(7, 4))
plt.scatter(study_hours, exam_scores, color='coral', alpha=0.7)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```

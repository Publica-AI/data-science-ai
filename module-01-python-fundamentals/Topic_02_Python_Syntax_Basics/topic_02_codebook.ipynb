## 2.1 Python Indentation

Python uses **indentation** (spaces at the start of a line) to group code into **blocks**.

- The standard is **4 spaces** per indentation level
- Every time a line ends with a colon `:`, the next line **must** be indented
- Indentation is **NOT optional** — it is a syntax rule in Python

**Why Python does this:** It forces readable, well-structured code. Most other languages use `{}` braces, but Python's approach means less visual clutter.

**Common Errors:**
- `TabError` — mixing tab characters and spaces
- `IndentationError` — inconsistent spacing within the same block


````python
# ============================================================
# CODE EXAMPLE 2.1a: Correct vs Incorrect Indentation
# ============================================================

# --- CORRECT: 4 spaces inside the if block ---
score = 85

if score >= 50:                          # colon → next line must be indented
    print("Result: Pass")                # 4 spaces — inside the if block
    print("Well done, keep it up!")      # still inside the if block

print("Assessment complete.")            # back to level 0 — always runs
````

````python
# ============================================================
# CODE EXAMPLE 2.1b: Nested indentation (multiple levels)
# ============================================================

# Each level adds 4 more spaces
temperature = 38
is_raining = True

if temperature > 30:               # Level 1 (4 spaces inside)
    print("It is hot outside.")    # 4 spaces
    if is_raining:                 # 4 spaces (another colon)
        print("And it is also raining!")  # 8 spaces (nested block)

print("End of weather check.")     # 0 spaces — always runs
````

## 2.2 Comments

A **comment** is a line in your code that Python completely ignores. It exists only for humans.

**Syntax:** Start the line with `#`

**When to use comments:**
- To explain **why** you made a particular choice (not what the code does — the code shows that)
- To describe the purpose of a section of code
- To temporarily disable a line of code for testing

**Good vs Bad Comments:**
```python
# ❌ Bad: just repeats the code
x = 5  # set x to 5

# ✅ Good: explains the WHY
interest_rate = 0.085  # 8.5% annual rate as per Central Bank policy
```


````python
# ============================================================
# CODE EXAMPLE 2.2a: Good commenting practices
# ============================================================

# --- Section: Calculate a student's final grade ---

# Weights based on course assessment structure
exam_score = 72         # out of 100
coursework_score = 88   # out of 100

# Exam counts 60%, coursework counts 40%
# This weighting was set by the academic committee
final_grade = (exam_score * 0.6) + (coursework_score * 0.4)

print("Final grade:", final_grade)
````

````python
# ============================================================
# CODE EXAMPLE 2.2b: Using comments to disable code (debugging)
# ============================================================

# During debugging, you can comment out lines to test different paths
name = "Kofi"

print("Hello,", name)                    # This runs
# print("Debugging: name value =", name)  # This is disabled (commented out)

# The commented line above would have run during testing
# We disabled it for the final version
````

## 2.3 The print() Function

The `print()` function is Python's primary way to display output.

**Three main uses:**

| Usage | Example | Output |
|-------|---------|--------|
| Print a string literal | `print("Hello")` | `Hello` |
| Print a variable | `print(name)` | value of name |
| Print multiple items | `print("Hi", name)` | `Hi Amara` |

**Key details:**
- When printing multiple items with commas, Python adds a **space** between them automatically
- You can print numbers without quotes: `print(42)`
- `print()` with nothing in it prints a **blank line**

**Common Mistake:** `print 'Hello'` — this is Python 2 syntax and causes a `SyntaxError` in Python 3


````python
# ============================================================
# CODE EXAMPLE 2.3a: All the ways to use print()
# ============================================================

student_name = "Abena"
student_age = 21
gpa = 3.75

# 1. Print string literals
print("=== Student Report ===")

# 2. Print a variable alone
print(student_name)

# 3. Print multiple items (comma-separated)
print("Name:", student_name)
print("Age:", student_age)
print("GPA:", gpa)

# 4. Print an empty line as a separator
print()

# 5. Print a number directly (no quotes needed for numbers)
print("Next year she will be:", student_age + 1)
````

````python
# ============================================================
# CODE EXAMPLE 2.3b: print() with sep and end parameters
# ============================================================

# Default: items separated by space
print("Python", "is", "fun")         # → Python is fun

# Custom separator
print("2024", "01", "15", sep="-")   # → 2024-01-15

# Custom end character (default is newline)
print("Loading", end="... ")          # stays on same line
print("Complete!")                    # → Loading... Complete!

print()  # blank line

# Practical use: printing a bar
for i in range(5):
    print("*", end="")  # print * with no newline
print()  # move to next line after the bar
````

## 2.4 The input() Function

The `input()` function **pauses** your program and waits for the user to type something and press Enter.

```python
user_response = input("Type your prompt here: ")
```

**The most important rule about input():**
> ⚠️ `input()` **ALWAYS returns a string** — even if the user types a number!

**Why this matters:**
```python
age = input("Enter your age: ")  # user types 25
# age = "25"  (a string, NOT the number 25)
```

If you try to add 5 to `"25"`, Python will join strings: `"25" + 5` → **TypeError!**

**Solution:** Convert with `int()` or `float()` (covered in Topic 3).


````python
# ============================================================
# CODE EXAMPLE 2.4a: Basic input() usage
# ============================================================

# The text inside input() is the prompt shown to the user
name = input("What is your name? ")

# Whatever the user typed is now stored in 'name' as a string
print("Hello,", name + "!")
print("Welcome to Python.")
````

````python
# ============================================================
# CODE EXAMPLE 2.4b: input() always returns a string — demo
# ============================================================

# Collect a number from the user
number_as_string = input("Enter a number: ")  # user types 42

# Check what type Python gives us
print("You entered:", number_as_string)
print("The type is:", type(number_as_string))  # → <class 'str'>

# This is why we need int() conversion before doing maths
# (We'll learn int() in Topic 3)
number_as_int = int(number_as_string)   # convert string → integer
print("After conversion, the type is:", type(number_as_int))  # → <class 'int'>
print("Add 10:", number_as_int + 10)
````

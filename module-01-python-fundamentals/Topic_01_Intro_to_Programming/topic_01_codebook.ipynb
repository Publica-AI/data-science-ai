## 1.1 What Programming Is

**Programming** is the process of writing precise, step-by-step instructions that a computer can follow to accomplish a task.

Think of it like writing a recipe:
- A recipe tells a cook *exactly* what to do, step by step
- A program tells a computer *exactly* what to do, step by step
- Computers are very literal — they do **exactly** what you write, nothing more

**Key ideas:**
- Programming is about **problem-solving** and **logical thinking**
- You do NOT need to be a math expert
- Like any skill, it improves with practice


````python
# ============================================================
# CODE EXAMPLE 1.1a: Your very first Python program
# This program gives Python the instruction to display text
# print() is a built-in function that outputs text to the screen
# ============================================================

print("Hello, World!")          # Display a greeting
print("I am learning Python!")  # Display a second message
print("Programming is fun.")    # Display a third message
````

````python
# ============================================================
# CODE EXAMPLE 1.1b: Programming as precise instructions
# This shows how Python follows each instruction one by one
# ============================================================

# Step 1: Store the user's name
student_name = "Amara"

# Step 2: Store the course name
course_name = "Python Programming"

# Step 3: Display a personalised welcome message
print("Welcome,", student_name)
print("You are enrolled in:", course_name)
print("Let's start coding!")
````

## 1.5 Running Python Scripts

The workflow for running a Python script in VS Code:

1. Create a new file and save it with a `.py` extension (e.g., `hello.py`)
2. Write your Python code
3. Open the integrated terminal: **Terminal → New Terminal**
4. Navigate to the correct folder: `cd path/to/your/folder`
5. Run the script: `python hello.py`

**Common Mistakes:**
- Being in the wrong directory → use `cd` to navigate
- Saving the file without `.py` extension
- Typos in the filename


````python
# ============================================================
# CODE EXAMPLE 1.5a: A complete introductory script
# This is the kind of script you would save as hello.py
# and run from the terminal
# ============================================================

# Display a welcome banner
print("=" * 40)
print("    Welcome to My First Python Script")
print("=" * 40)

# Ask the user for their name
name = input("Please enter your name: ")

# Ask the user for their age
age = input("Please enter your age: ")

# Display a personalised message
print("\nHello,", name + "!")
print("You are", age, "years old.")
print("Welcome to Python Programming!")
````

## 1.6 Python REPL vs Script Files

Python offers two ways to run code:

### The REPL (Interactive Mode)
- Access by typing `python` in the terminal
- Type one line → press Enter → see immediate result
- Great for: quick experiments, testing ideas, calculator-style use
- Downside: code is lost when you close the terminal

### Script Files (.py)
- Write code in a `.py` file and run with `python filename.py`
- Great for: real programs, multi-line code, code you want to save
- Code is permanently saved

**Rule of thumb:** Use the REPL to explore; use scripts to build.


````python
# ============================================================
# CODE EXAMPLE 1.6a: REPL-style quick experiments
# These are the kinds of things you would test in the REPL
# ============================================================

# Quick maths
print(2 + 2)            # → 4
print(10 / 3)           # → 3.3333...
print(10 // 3)          # → 3  (floor division)
print(10 % 3)           # → 1  (remainder)

# Quick string test
print("hello".upper())  # → HELLO
print(len("Python"))    # → 6 (number of characters)
````

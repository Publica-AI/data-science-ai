## 13.1 Opening and Closing Files

- `open(filename, mode, encoding='utf-8')` returns a file object
- File modes: `'r'` read (default), `'w'` write/overwrite, `'a'` append, `'x'` create new
- Always call `.close()` after working with the file — or better, use `with`
- Always specify `encoding='utf-8'` for consistent cross-platform behavior
- `FileNotFoundError` is raised if you open a non-existent file in `'r'` mode


```python
# CODE EXAMPLE 13.1a: File modes

import os

# Write a file (creates it if it doesn't exist)
f = open('example.txt', 'w', encoding='utf-8')
f.write('Hello, file!\n')
f.close()

print('File exists:', os.path.exists('example.txt'))   # True

# Read the file
f = open('example.txt', 'r', encoding='utf-8')
print(f.read())
f.close()

# Append to the file
f = open('example.txt', 'a', encoding='utf-8')
f.write('Second line.\n')
f.close()
```

```python
# CODE EXAMPLE 13.1b: Handling FileNotFoundError

try:
    f = open('missing_file.txt', 'r')
    content = f.read()
    f.close()
except FileNotFoundError as e:
    print(f'Error: {e}')

# x mode — fails if file already exists
try:
    f = open('example.txt', 'x')  # file already exists
except FileExistsError as e:
    print(f'File already exists: {e}')
```

## 13.2 Reading from Files

- `f.read()` — reads entire file as one string
- `f.readline()` — reads one line at a time (including the `\n`)
- `f.readlines()` — reads all lines and returns a list of strings
- `for line in f:` — most memory-efficient; reads line by line
- Use `.strip()` to remove leading/trailing whitespace and `\n`


```python
# CODE EXAMPLE 13.2a: Three ways to read

# First create a test file
with open('test_read.txt', 'w', encoding='utf-8') as f:
    f.write('Line one\nLine two\nLine three\n')

# read() — entire content as a string
with open('test_read.txt', 'r', encoding='utf-8') as f:
    content = f.read()
    print('read():')
    print(content)

# readlines() — list of lines
with open('test_read.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    print('readlines():', lines)

# Iterate line by line
with open('test_read.txt', 'r', encoding='utf-8') as f:
    print('for loop:')
    for line in f:
        print(line.strip())   # strip removes \n
```

```python
# CODE EXAMPLE 13.2b: Processing file contents

# Create a scores file
with open('scores.txt', 'w', encoding='utf-8') as f:
    f.write('Amara 85\nKofi 72\nAma 91\nYaw 65\n')

# Read and process
total = 0
count = 0
with open('scores.txt', 'r', encoding='utf-8') as f:
    for line in f:
        parts = line.strip().split()
        name, score = parts[0], int(parts[1])
        print(f'{name}: {score}')
        total += score
        count += 1

print(f'Average: {total / count:.1f}')
```

## 13.3 Writing to Files

- `f.write(string)` — writes a string; does NOT add newline automatically
- `f.writelines(list)` — writes each string in the list (no auto-newlines)
- Mode `'w'` — overwrites the file from scratch
- Mode `'a'` — appends to the end of existing content
- Always add `'\n'` manually to separate lines


```python
# CODE EXAMPLE 13.3a: write() and writelines()

# write() — single strings
with open('output.txt', 'w', encoding='utf-8') as f:
    f.write('First line\n')
    f.write('Second line\n')
    f.write('Third line\n')

# writelines() — list of strings
lines = ['Alice,85\n', 'Bob,72\n', 'Carol,91\n']
with open('results.txt', 'w', encoding='utf-8') as f:
    f.writelines(lines)

# Verify both files
with open('output.txt') as f:
    print(f.read())
with open('results.txt') as f:
    print(f.read())
```

```python
# CODE EXAMPLE 13.3b: Append mode

# Start fresh
with open('log.txt', 'w', encoding='utf-8') as f:
    f.write('Session started\n')

# Append entries (simulates logging)
import datetime
events = ['User logged in', 'Viewed dashboard', 'Updated profile']
with open('log.txt', 'a', encoding='utf-8') as f:
    for event in events:
        timestamp = datetime.datetime.now().strftime('%H:%M:%S')
        f.write(f'[{timestamp}] {event}\n')

# Read the full log
with open('log.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

## 13.4 The with Statement

- `with open(...) as f:` — context manager syntax
- Automatically closes the file when the block ends, even if an exception occurs
- Eliminates the risk of forgetting `.close()` or missing it due to an error
- Can open multiple files in one with statement
- Best practice: always use `with open()` instead of manual open/close


```python
# CODE EXAMPLE 13.4a: with vs manual open/close

# WITHOUT with — risky
f = open('data.txt', 'w')
f.write('test')
# If an error happened here, f.close() would never run
f.close()

# WITH with — safe and clean
with open('data.txt', 'w', encoding='utf-8') as f:
    f.write('test')
# f is automatically closed here, guaranteed

# Verify the file is closed
print('File closed:', f.closed)   # True

# with for reading
with open('data.txt', 'r', encoding='utf-8') as f:
    content = f.read()
print(content)   # can use content here — file was auto-closed above
```

```python
# CODE EXAMPLE 13.4b: Multiple files in one with

# Create input file
with open('input.txt', 'w', encoding='utf-8') as f:
    f.write('apple\nbanana\nmango\n')

# Copy + transform: read one file, write to another — both open at once
with open('input.txt', 'r', encoding='utf-8') as src, \
     open('upper.txt', 'w', encoding='utf-8') as dst:
    for line in src:
        dst.write(line.upper())

with open('upper.txt', 'r') as f:
    print(f.read())
```

## 13.5 Working with CSV Files

CSV = Comma-Separated Values — a universal format for tables.

- `csv.reader(f)` — reads rows as lists
- `csv.writer(f)` — writes rows from lists; `writerow()` writes one row
- `csv.DictReader(f)` — rows as dicts with column names as keys
- `csv.DictWriter(f, fieldnames=[...])` — writes dicts as rows
- Always open CSV files with `newline=''` to prevent duplicate blank lines on Windows


```python
# CODE EXAMPLE 13.5a: csv.writer and csv.reader

import csv

# Write CSV
with open('grades.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Name', 'Score', 'Grade'])   # header
    writer.writerow(['Amara', 85, 'A'])
    writer.writerow(['Kofi',  72, 'B'])
    writer.writerow(['Ama',   91, 'A'])

# Read CSV
with open('grades.csv', 'r', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
# CODE EXAMPLE 13.5b: DictReader and DictWriter

import csv

students = [
    {'name': 'Amara', 'score': 85, 'grade': 'A'},
    {'name': 'Kofi',  'score': 72, 'grade': 'B'},
    {'name': 'Ama',   'score': 91, 'grade': 'A'},
]

# Write with DictWriter
with open('students.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['name', 'score', 'grade'])
    writer.writeheader()
    writer.writerows(students)

# Read with DictReader — access by column name
with open('students.csv', 'r', newline='', encoding='utf-8') as f:
    for row in csv.DictReader(f):
        print(f"{row['name']:10} scored {row['score']} — Grade {row['grade']}")
```

## Topic 13 Recap

**Open:** `open('file.txt', 'r', encoding='utf-8')` — modes: r, w, a, x.

**Read:** `read()` all | `readline()` one | `for line in f:` most efficient.

**Write:** `write('text\n')` | `writelines(list)` — add `\n` manually.

**with:** `with open(...) as f:` — auto-closes; always use this.

**CSV:** `csv.writer` / `csv.reader` for lists; `DictWriter` / `DictReader` for named columns.

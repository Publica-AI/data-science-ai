## 11.1 Types of Errors

Three categories:
- **Syntax error** — broken grammar; caught before running (missing colon, unmatched bracket)
- **Runtime error** — valid syntax but crashes during execution (IndexError, ValueError, ZeroDivisionError)
- **Logic error** — runs without crashing but produces wrong output; hardest to find


```python
# CODE EXAMPLE 11.1a: The three types of errors

# --- Syntax Error (uncomment to see) ---
# def greet(:          # SyntaxError: invalid syntax
#     print('hello')

# --- Runtime Errors ---
try:
    print(int('hello'))    # ValueError
except ValueError as e:
    print(f'RuntimeError caught: {e}')

try:
    result = 10 / 0        # ZeroDivisionError
except ZeroDivisionError as e:
    print(f'RuntimeError caught: {e}')

# --- Logic Error (no crash, wrong result) ---
def calculate_average(nums):
    return sum(nums) / 10    # BUG: should be len(nums)

print(calculate_average([5, 10, 15]))  # 3.0 (wrong: should be 10.0)
```

```python
# CODE EXAMPLE 11.1b: Common runtime error types

errors = [
    (lambda: [][5], 'IndexError'),
    (lambda: int('abc'), 'ValueError'),
    (lambda: 'a' + 1, 'TypeError'),
    (lambda: {'k': 1}['z'], 'KeyError'),
]

for func, name in errors:
    try:
        func()
    except Exception as e:
        print(f'{name}: {e}')
```

## 11.2 try / except Blocks

- Put risky code in the `try` block
- If an error occurs, execution jumps to the matching `except` block
- Always catch **specific** exception types — avoid bare `except:`
- Use `as e` to capture the error message
- Multiple except clauses can catch different error types


```python
# CODE EXAMPLE 11.2a: try/except with specific exceptions

def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print('Error: cannot divide by zero.')
        return None
    except TypeError:
        print('Error: both arguments must be numbers.')
        return None
    return result

print(safe_divide(10, 2))    # 5.0
print(safe_divide(10, 0))    # Error: cannot divide by zero.
print(safe_divide(10, 'a'))  # Error: both arguments must be numbers.
```

```python
# CODE EXAMPLE 11.2b: Catching the exception message with 'as e'

def get_integer_input(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError as e:
            print(f'Invalid input ({e}). Please enter a whole number.')

# Uncomment to run interactively:
# age = get_integer_input('Enter your age: ')
# print(f'You are {age} years old.')

# Simulated test:
try:
    int('abc')
except ValueError as e:
    print(f'Caught ValueError: {e}')
```

## 11.3 else and finally Clauses

- `else` — runs only if the `try` block succeeded (no exception)
- `finally` — runs ALWAYS, regardless of whether an exception occurred
- Use `else` for code that should only run on success
- Use `finally` for cleanup: closing files, printing closing messages, releasing resources


```python
# CODE EXAMPLE 11.3a: else and finally in action

def parse_and_double(text):
    try:
        number = int(text)
    except ValueError:
        print(f'Could not convert "{text}" to integer.')
    else:
        result = number * 2
        print(f'Success: {number} doubled is {result}')  # only on success
    finally:
        print('Attempt finished.')    # always runs

parse_and_double('5')      # runs else + finally
print()
parse_and_double('abc')    # runs except + finally
```

```python
# CODE EXAMPLE 11.3b: finally for cleanup simulation

def read_data(filename):
    print(f'Opening {filename}...')
    try:
        # Simulate reading
        if filename == 'missing.txt':
            raise FileNotFoundError(f'{filename} not found.')
        data = 'file contents'
        return data
    except FileNotFoundError as e:
        print(f'Error: {e}')
        return None
    finally:
        print(f'Closing {filename}.')   # always runs

result = read_data('data.txt')
print('Result:', result)
print()
result = read_data('missing.txt')
print('Result:', result)
```

## 11.4 Raising Exceptions

- Use `raise` to trigger an exception intentionally
- Raise when user input violates a business rule Python would not catch on its own
- Always include a descriptive message: `raise ValueError('message')`
- Validate at the boundary — raise early before bad data propagates
- Common types to raise: `ValueError`, `TypeError`, `AttributeError`


```python
# CODE EXAMPLE 11.4a: Raising ValueError

def set_score(score):
    if not isinstance(score, (int, float)):
        raise TypeError(f'Score must be a number, got {type(score).__name__}.')
    if score < 0 or score > 100:
        raise ValueError(f'Score {score} is out of range (0–100).')
    return score

# Valid
print(set_score(85))       # 85

# Invalid — caught by caller
try:
    set_score(150)
except ValueError as e:
    print(f'Caught: {e}')

try:
    set_score('A+')
except TypeError as e:
    print(f'Caught: {e}')
```

```python
# CODE EXAMPLE 11.4b: Raise used in a function pipeline

def validate_name(name):
    if not isinstance(name, str) or len(name.strip()) == 0:
        raise ValueError('Name must be a non-empty string.')
    return name.strip().title()

def register_student(name, age):
    name = validate_name(name)
    if not isinstance(age, int) or age < 5 or age > 100:
        raise ValueError(f'Age {age} is not valid for a student.')
    print(f'Registered: {name}, age {age}')

try:
    register_student('  amara  ', 20)   # valid
    register_student('', 20)            # raises ValueError
except ValueError as e:
    print(f'Registration failed: {e}')
```

## 11.5 Debugging Techniques

- **Print debugging** — insert temporary print statements to track variable values
- **Tracebacks** — Python's error report; read from the bottom up to find the root cause
- **assert** — checks a condition and raises AssertionError if false; good for catching programmer assumptions
- Remove debug prints before finalizing code


```python
# CODE EXAMPLE 11.5a: Print debugging

def calculate_final_grade(scores):
    print(f'[DEBUG] scores received: {scores}')   # trace input
    total = sum(scores)
    print(f'[DEBUG] total: {total}')              # trace intermediate
    average = total / len(scores)
    print(f'[DEBUG] average: {average}')          # trace output
    return average

grade = calculate_final_grade([70, 85, 90, 60])
print(f'Final grade: {grade}')
```

```python
# CODE EXAMPLE 11.5b: assert statements

def divide(a, b):
    assert b != 0, 'Denominator must not be zero'
    assert isinstance(a, (int, float)), f'Expected number, got {type(a).__name__}'
    return a / b

print(divide(10, 2))   # 5.0

try:
    divide(10, 0)
except AssertionError as e:
    print(f'AssertionError: {e}')

# assert is useful for checking function preconditions during development
# Remove or disable assertions in production code if performance matters
```

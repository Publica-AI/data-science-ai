## 8.1 Creating and Accessing Lists

A **list** is a mutable, ordered collection of items.

- Created with square brackets: `[item1, item2, item3]`
- Zero-based indexing: first item is `list[0]`
- Negative indexing: `list[-1]` is the last item
- Slicing works the same as strings: `list[start:stop]`
- `len(list)` gives the number of items
- `item in list` checks membership


```python
# CODE EXAMPLE 8.1a: Creating and accessing lists

# Create lists
countries = ["Ghana", "Nigeria", "Kenya", "Egypt", "South Africa"]
scores = [82, 91, 74, 88, 95]

# Access by index
print(countries[0])      # Ghana (first)
print(countries[-1])     # South Africa (last)
print(countries[1:3])    # ['Nigeria', 'Kenya'] (slice)

# Length and membership
print(len(countries))         # 5
print("Kenya" in countries)   # True
print("Japan" in countries)   # False
```

```python
# CODE EXAMPLE 8.1b: List with mixed types and nested lists

# Lists can hold any type
student_record = ["Amara", 22, 3.75, True]
name, age, gpa, is_enrolled = student_record   # unpacking
print(f"{name}, age {age}, GPA {gpa}")

# Lists can be nested (list of lists)
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[1][2])   # 6 (row 1, column 2)
```

## 8.2 Modifying Lists

Lists are **mutable** — you can change them in place.

- `.append(item)` — adds to the end
- `.insert(index, item)` — inserts at a position
- `.pop(index)` — removes at index and **returns** the item (default: last item)
- `.remove(value)` — removes the first occurrence of a value

**Common confusion:** `.remove()` takes a VALUE, `.pop()` takes an INDEX.


```python
# CODE EXAMPLE 8.2a: Modifying lists

cart = ["bread", "milk", "eggs"]

# append — add to end
cart.append("butter")
print("After append:", cart)

# insert — add at position
cart.insert(1, "juice")       # insert at index 1
print("After insert:", cart)

# pop — remove at index (returns it)
removed = cart.pop(0)         # remove first item
print(f"Removed: {removed}, Cart: {cart}")

# remove — remove by value
cart.remove("milk")           # remove by value
print("After remove:", cart)
```

```python
# CODE EXAMPLE 8.2b: Additional list operations

numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# Sort in place
numbers.sort()
print("Sorted:", numbers)

# Sort descending
numbers.sort(reverse=True)
print("Descending:", numbers)

# Reverse in place
numbers.reverse()
print("Reversed:", numbers)

# Count and index
print("Count of 1:", numbers.count(1))
print("First 1 is at index:", numbers.index(1))
```

## 8.3 Iterating Over Lists

- `for item in list:` — iterates over each value
- `for i, item in enumerate(list):` — gives index + value
- **Never modify** a list while iterating over it — it causes bugs


```python
# CODE EXAMPLE 8.3a: Iterating with enumerate

students = ["Amara", "Kofi", "Ama", "Kweku"]

# Basic for loop
print("Class list:")
for student in students:
    print(" -", student)

print()

# With enumerate (number from 1)
print("Numbered list:")
for index, student in enumerate(students, start=1):
    print(f"  {index}. {student}")
```

## 8.4 Tuples and Immutability

A **tuple** is an immutable, ordered collection — like a read-only list.

- Created with parentheses: `(item1, item2)`
- Accessing works exactly like lists: `t[0]`, `t[-1]`, slicing
- Cannot add, remove, or modify items after creation
- Single-item tuple requires a trailing comma: `(42,)` not `(42)`
- Can be used as dictionary keys (lists cannot)
- **Tuple unpacking:** `a, b, c = (1, 2, 3)`


```python
# CODE EXAMPLE 8.4a: Tuples

# Fixed data — good for tuples
coordinates = (5.6037, -0.1870)    # Accra GPS coordinates
rgb = (255, 165, 0)                # Orange in RGB

# Access like a list
print(coordinates[0])              # 5.6037
print(rgb[-1])                     # 0

# Tuple unpacking
lat, lon = coordinates
print(f"Latitude: {lat}, Longitude: {lon}")

r, g, b = rgb
print(f"R={r}, G={g}, B={b}")

# Single-item tuple
single = (42,)         # ← comma required!
print(type(single))    # <class 'tuple'>
not_tuple = (42)
print(type(not_tuple)) # <class 'int'>
```

## Topic 8 Recap

**List:** `[a, b, c]` — mutable, ordered, zero-indexed.

**Modification:** `.append()` (end), `.insert(i, v)` (at position), `.pop(i)` (remove+return by index), `.remove(v)` (remove by value).

**Iteration:** `for item in list:` — `for i, item in enumerate(list):` for index+value.

**Tuple:** `(a, b, c)` — immutable, ordered; single item needs comma `(x,)`; use for fixed data.

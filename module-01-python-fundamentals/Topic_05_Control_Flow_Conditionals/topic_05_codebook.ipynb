## 5.1 The if Statement

The `if` statement runs a block of code only when a condition evaluates to `True`.

```
if condition:
    # this block runs only when condition is True
```

- The condition is any expression that returns `True` or `False`
- The colon `:` at the end is **mandatory**
- The indented block (4 spaces) only executes when condition is `True`
- Code after the block (at the original indent level) always runs


````python
# CODE EXAMPLE 5.1a: Basic if statement

age = 20

if age >= 18:
    print("Access granted.")     # runs because 20 >= 18 is True
    print("Welcome to the system.")

print("Check complete.")         # always runs
````

````python
# CODE EXAMPLE 5.1b: if with logical operators

temperature = 38
is_headache = True

# Combining conditions with 'and'
if temperature >= 38 and is_headache:
    print("Symptoms suggest fever.")
    print("Please see a doctor.")

# if with 'not'
stock_available = False
if not stock_available:
    print("Item out of stock — notify supplier.")
````

## 5.2 The if-else Statement

The `if-else` gives you exactly two paths. One will ALWAYS execute.

- If condition is `True` → `if` block runs
- If condition is `False` → `else` block runs
- **`else` never has a condition** — just `else:`


````python
# CODE EXAMPLE 5.2a: if-else

score = float(input("Enter your exam score: "))

if score >= 50:
    print("Result: PASS")
    print("Well done!")
else:
    print("Result: FAIL")
    print("Please prepare for the resit.")
````

````python
# CODE EXAMPLE 5.2b: Using if-else to assign a value

is_weekend = True

# Assign different values based on condition
if is_weekend:
    schedule = "Leisure time"
else:
    schedule = "Work day"

print("Today's schedule:", schedule)
````

## 5.3 The if-elif-else Statement

`elif` means "else if" — it adds extra conditions to check.

- Python checks conditions **top to bottom**
- The **first** `True` condition executes and the rest are **skipped**
- `else` is the catch-all for nothing matching above
- Use `elif` for mutually exclusive options — NOT a series of `if`s


````python
# CODE EXAMPLE 5.3a: Grade classification with if-elif-else

score = float(input("Enter score: "))

if score >= 90:
    grade = "A"
    remark = "Distinction"
elif score >= 80:
    grade = "B"
    remark = "Merit"
elif score >= 70:
    grade = "C"
    remark = "Credit"
elif score >= 60:
    grade = "D"
    remark = "Pass"
else:
    grade = "F"
    remark = "Fail"

print(f"Grade: {grade} — {remark}")
````

````python
# CODE EXAMPLE 5.3b: Loan eligibility checker

income = float(input("Monthly income (GHS): "))
credit_score = int(input("Credit score (300-850): "))

if income >= 5000 and credit_score >= 700:
    print("Approved for Premium Loan")
elif income >= 3000 and credit_score >= 600:
    print("Approved for Standard Loan")
elif income >= 1500 and credit_score >= 500:
    print("Approved for Micro Loan")
else:
    print("Application not approved at this time.")
````

## 5.4 Truthy and Falsy Values

In Python, any value can be used as a condition. Values that behave like `False` are **falsy**:

- `0`, `0.0` — zero numbers
- `""` — empty string
- `[]`, `{}`, `set()` — empty collections
- `None`

**Everything else is truthy.** This allows concise checks:

- `if my_list:` instead of `if len(my_list) > 0:`
- `if name:` instead of `if name != ""`


````python
# CODE EXAMPLE 5.4a: Truthy and falsy demonstration

# Falsy values
print(bool(0))      # False
print(bool(""))     # False
print(bool([]))     # False
print(bool(None))   # False

# Truthy values
print(bool(1))      # True
print(bool("Hi"))   # True
print(bool([1,2]))  # True

# Concise usage
cart = []
if not cart:
    print("Your cart is empty — add some items!")
````

````python
# CODE EXAMPLE 5.4b: Practical truthy/falsy patterns

name = input("Enter your name (press Enter to skip): ")

# Truthy check — name is truthy if non-empty
if name:
    print(f"Welcome, {name}!")
else:
    print("Welcome, anonymous user!")

# Number truthy check
balance = 0
if balance:
    print(f"Balance: GHS {balance}")
else:
    print("Account balance is zero.")
````

## Topic 5 Recap

**`if condition:`** — block runs only when `True`. Colon required. 4-space indent.

**`if ... else:`** — exactly two paths; one always runs; `else` has no condition.

**`if ... elif ... else:`** — first matching condition wins; use `elif` not separate `if`s for mutually exclusive cases.

**Truthy/falsy:** `0`, `""`, `[]`, `None` are falsy. Use `if my_list:` instead of `if len(my_list) > 0:`

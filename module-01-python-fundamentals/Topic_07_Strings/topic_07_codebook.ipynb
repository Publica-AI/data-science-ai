## 7.1 String Indexing and Slicing

**Indexing** accesses one character at a specific position (zero-based):
- `s[0]` — first character
- `s[-1]` — last character (negative counts from the end)

**Slicing** extracts a substring:
- `s[start:stop]` — from start up to (not including) stop
- `s[start:stop:step]` — with a step
- `s[::-1]` — reverses the string


````python
# CODE EXAMPLE 7.1a: Indexing and slicing

language = "Python"

# Positive indexing (left to right, starts at 0)
print(language[0])      # P
print(language[2])      # t
print(language[5])      # n (last character, index = len-1)

# Negative indexing (right to left, starts at -1)
print(language[-1])     # n (last)
print(language[-3])     # h (third from end)

# Slicing
print(language[0:3])    # Pyt (stop 3 is excluded)
print(language[2:])     # thon (from index 2 to end)
print(language[:4])     # Pyth (from start to index 3)
print(language[::-1])   # nohtyP (reversed)
````

````python
# CODE EXAMPLE 7.1b: Practical slicing

email = "amara.mensah@gmail.com"

# Extract username (everything before @)
at_position = email.find("@")
username = email[:at_position]
print("Username:", username)    # amara.mensah

# Extract domain (everything after @)
domain = email[at_position + 1:]
print("Domain:", domain)        # gmail.com

# Check if palindrome
word = "racecar"
print(f"{word} is palindrome: {word == word[::-1]}")
````

## 7.2 String Immutability

Strings are **immutable** — you cannot change individual characters after creation.

```python
name = "python"
name[0] = "P"   # TypeError! Cannot modify in place
```

To "change" a string, create a **new** one:
- Use string methods (they return new strings)
- Use slicing and concatenation

**Critical:** When you call a method like `.strip()`, you must reassign the result or it is lost.


````python
# CODE EXAMPLE 7.2a: Working around immutability

original = "hello"

# Method 1: slicing + concatenation
capitalized = "H" + original[1:]    # 'Hello'
print(capitalized)

# Method 2: use a method and reassign
city = "  kumasi  "
city = city.strip()         # reassign to keep the result
print(city)                 # 'kumasi'

# WRONG — strip() is called but result discarded
city2 = "  accra  "
city2.strip()               # does nothing useful!
print(city2)                # still '  accra  '
````

## 7.3 Common String Methods

All string methods return a **new** string — they do not modify the original.

- `.lower()` / `.upper()` — change case
- `.strip()` / `.lstrip()` / `.rstrip()` — remove whitespace
- `.replace(old, new)` — substitute text
- `.startswith(sub)` / `.endswith(sub)` — check prefix/suffix
- `.split(sep)` — split into a list
- `sep.join(list)` — join list into string
- `.find(sub)` — find index of first occurrence (-1 if not found)
- `.count(sub)` — count occurrences


````python
# CODE EXAMPLE 7.3a: String methods in action

raw_input = "  Hello, Python World!  "

# Clean and transform
cleaned = raw_input.strip()
lowered = cleaned.lower()
upper   = cleaned.upper()
replaced = cleaned.replace("Python", "Amazing")

print(cleaned)    # 'Hello, Python World!'
print(lowered)    # 'hello, python world!'
print(upper)      # 'HELLO, PYTHON WORLD!'
print(replaced)   # 'Hello, Amazing World!'

# Check and search
print(cleaned.startswith("Hello"))   # True
print(cleaned.endswith("?"))         # False
print(cleaned.find("Python"))        # 7 (index where 'Python' starts)
print(cleaned.count("l"))            # 3
````

````python
# CODE EXAMPLE 7.3b: split() and join()

sentence = "Python Data Science Machine Learning"

# split() breaks string into list of words
words = sentence.split()                  # split on whitespace by default
print(words)                              # ['Python', 'Data', ...]
print("Word count:", len(words))          # 4

# CSV split on comma
csv_line = "Amara,22,Ghana,Data Science"
fields = csv_line.split(",")
print(fields)                             # ['Amara', '22', 'Ghana', 'Data Science']

# join() brings a list back into a string
dashes = " - ".join(words)
print(dashes)                             # Python - Data - Science - Machine - Learning
````

## 7.4 String Formatting — f-Strings

f-strings (formatted string literals) embed Python expressions directly inside strings.

```python
name = "Amara"
print(f"Hello, {name}!")    # Hello, Amara!
```

- Prefix the string with `f` or `F`
- Put any Python expression inside `{}`
- Format specifiers: `{value:.2f}` for 2 decimals, `{value:>10}` for right-align
- Introduced in Python 3.6 — the modern standard


````python
# CODE EXAMPLE 7.4a: f-string features

name = "Kofi"
score = 87.567
rank = 3

# Basic embedding
print(f"Student: {name}")

# Expressions inside {}
print(f"Score + bonus: {score + 5:.1f}")

# Format decimal places
print(f"Score: {score:.2f}")          # 87.57

# Padding/alignment
print(f"Rank: {rank:02d}")            # Rank: 03 (zero-padded)

# Method calls inside {}
print(f"Name uppercase: {name.upper()}")

# Multi-line report
print(f"""
--- Student Report ---
Name  : {name}
Score : {score:.1f}
Grade : {'A' if score >= 80 else 'B'}
""")
````

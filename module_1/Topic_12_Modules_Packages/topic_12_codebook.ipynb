## 12.1 What Are Modules

A **module** is a Python file (`.py`) containing functions, variables, and classes.

- Modules group related code so it can be reused across many programs
- Python's Standard Library is a collection of modules that ship with Python
- Every `.py` file you write is itself a module — importable by others
- `import module_name` loads the module into your program


```python
# CODE EXAMPLE 12.1a: Using the math module

import math

print(math.pi)               # 3.141592653589793
print(math.e)                # 2.718281828...
print(math.sqrt(144))        # 12.0
print(math.factorial(5))     # 120
print(math.log(100, 10))     # 2.0  (log base 10 of 100)
print(math.floor(4.7))       # 4
print(math.ceil(4.2))        # 5
```

```python
# CODE EXAMPLE 12.1b: Exploring module contents with dir()

import math

# dir() lists all names in a module
print([name for name in dir(math) if not name.startswith('_')])

# help() shows documentation
# help(math.sqrt)   # uncomment to see

# Check if something is in a module
print(hasattr(math, 'tau'))   # True in Python 3.6+
print(math.tau)               # 2 * pi
```

## 12.2 Importing Modules

Three import styles:

- `import module` — import whole module; access as `module.name`
- `from module import name` — import specific item; use directly
- `import module as alias` — rename for convenience

Avoid `from module import *` — it pollutes the namespace and makes code harder to read.


```python
# CODE EXAMPLE 12.2a: Three import styles

# Style 1: full module import
import math
print(math.sqrt(36))         # 6.0

# Style 2: specific import
from math import sqrt, pi
print(sqrt(36))              # 6.0 — no prefix needed
print(pi)                    # 3.14159...

# Style 3: alias
import math as m
print(m.sqrt(36))            # 6.0

import datetime as dt
print(dt.date.today())       # today's date
```

```python
# CODE EXAMPLE 12.2b: Importing from random

import random

# Random integer between 1 and 10 (inclusive)
print(random.randint(1, 10))

# Random float between 0 and 1
print(random.random())

# Random choice from a list
fruits = ['apple', 'banana', 'mango', 'orange']
print(random.choice(fruits))

# Shuffle a list in-place
random.shuffle(fruits)
print(fruits)

# Sample without replacement
print(random.sample(fruits, 2))
```

## 12.3 Creating Your Own Module

- Write functions and variables in a `.py` file
- Save it in the same directory as your main script
- Import it with `import filename` (without the .py extension)
- Access functions with `module.function_name()`
- Modules can also contain constants and classes


```python
# CODE EXAMPLE 12.3a: Simulating a custom module
# In a real project you would save this to converters.py and import it.
# Here we define the functions inline to demonstrate the concept.

# --- Contents of converters.py ---
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

def km_to_miles(km):
    return km * 0.621371

def kg_to_pounds(kg):
    return kg * 2.20462

EARTH_RADIUS_KM = 6371

# --- In main.py you would write: import converters ---
# Using the functions:
print(celsius_to_fahrenheit(100))   # 212.0
print(km_to_miles(10))              # 6.21371
print(kg_to_pounds(70))             # 154.32
print(EARTH_RADIUS_KM)              # 6371
```

```python
# CODE EXAMPLE 12.3b: The __name__ == '__main__' guard
# When a module is imported, all its top-level code runs.
# The guard prevents test/demo code from running on import.

# --- Contents of mymodule.py ---
def greet(name):
    return f'Hello, {name}!'

def add(a, b):
    return a + b

# This block ONLY runs when mymodule.py is run directly:
if __name__ == '__main__':
    print('Running module tests...')
    print(greet('World'))     # Hello, World!
    print(add(3, 4))          # 7

# When imported elsewhere, the if block is skipped
print('Demonstrating __name__:', __name__)
```

## 12.4 The Standard Library

Key Standard Library modules:

- `math` — mathematical functions and constants
- `random` — random number generation and selection
- `datetime` — date and time operations
- `os` — operating system interface (files, directories, environment)
- `sys` — Python interpreter information
- `string` — string constants (letters, digits, punctuation)


```python
# CODE EXAMPLE 12.4a: datetime module

import datetime

# Current date and time
now = datetime.datetime.now()
print('Now:', now)
print('Year:', now.year)
print('Month:', now.month)
print('Day:', now.day)
print('Weekday:', now.strftime('%A'))   # Monday, Tuesday...

# Date arithmetic
today = datetime.date.today()
birthday = datetime.date(today.year, 12, 25)
days_until = (birthday - today).days
print(f'Days until Dec 25: {days_until}')

# Formatting
print(today.strftime('%d/%m/%Y'))
```

```python
# CODE EXAMPLE 12.4b: os and sys modules

import os
import sys

# os — operating system
print('Current directory:', os.getcwd())
print('Files here:', os.listdir('.')[:5])   # first 5 files
print('Path separator:', os.sep)           # \ on Windows, / on Unix

# os.path — file path utilities
path = '/home/user/documents/report.txt'
print('Base name:', os.path.basename(path))   # report.txt
print('Directory:', os.path.dirname(path))    # /home/user/documents

# sys — Python interpreter
print('Python version:', sys.version[:6])
print('Platform:', sys.platform)
```

## 12.5 Installing and Using Packages

- PyPI (Python Package Index) hosts 400,000+ packages
- Install with: `pip install package-name` in the terminal
- Import like any module after installation
- Common packages: `requests` (HTTP), `pandas` (data), `numpy` (numbers), `colorama` (colors)
- List installed packages: `pip list`
- Uninstall: `pip uninstall package-name`


```python
# CODE EXAMPLE 12.5a: Using colorama (pip install colorama)
# If not installed, the try/except handles it gracefully

try:
    from colorama import Fore, Style, init
    init()  # needed on Windows
    print(Fore.GREEN + 'SUCCESS: colorama is installed!')
    print(Fore.YELLOW + 'WARNING: This is a yellow warning.')
    print(Fore.RED + 'ERROR: This is a red error message.')
    print(Style.RESET_ALL + 'Back to normal text.')
except ImportError:
    print('colorama not installed. Run: pip install colorama')
    print('Without colorama, output shows without color.')
```

```python
# CODE EXAMPLE 12.5b: Using requests (pip install requests)
# requests makes HTTP calls to web APIs

try:
    import requests
    response = requests.get('https://httpbin.org/get', timeout=5)
    print('Status code:', response.status_code)
    print('Response type:', type(response.json()))
except ImportError:
    print('requests not installed. Run: pip install requests')
except Exception as e:
    print(f'Network error: {e}')

# Common HTTP status codes:
# 200 = OK, 404 = Not Found, 500 = Server Error
```

## 10.1 Defining and Calling Functions

A **function** is a named, reusable block of code.

Key ideas:
- Use `def` keyword followed by the function name, parentheses, and a colon
- The indented block is the **function body** — it runs only when called
- **Defining** a function does NOT run it — you must **call** it
- Call a function by writing its name followed by `()`
- You can call the same function as many times as you want — this is the **DRY principle** (Don't Repeat Yourself)


```python
# CODE EXAMPLE 10.1a: Defining and calling a function

# Define the function — nothing runs yet
def greet():
    print("Hello, welcome to Python!")
    print("Let's write some functions.")

# Call the function — NOW it runs
greet()
greet()    # can call multiple times!

# Without a call — this function body never runs
def never_called():
    print("I will never run unless called.")
```

```python
# CODE EXAMPLE 10.1b: DRY — replacing repeated code with functions

# WITHOUT functions — repeated code
print("--- Report ---")
print("Student: Amara")
print("--- Report ---")
print("Student: Kofi")

# WITH functions — DRY
def print_report_header():
    print("--- Report ---")

print_report_header()
print("Student: Amara")
print_report_header()
print("Student: Kofi")

# Now if the header format changes, you update ONE place only
```

## 10.2 Parameters and Arguments

Functions become truly powerful when they work with **different data** each call.

- **Parameter** — the placeholder name in the function **definition**
- **Argument** — the actual value passed in the **call**
- Order matters — arguments are matched to parameters **by position**
- You can have multiple parameters separated by commas


```python
# CODE EXAMPLE 10.2a: Single parameter

# 'name' is the PARAMETER — a placeholder
def greet(name):
    print(f"Hello, {name}!")

# "Amara" and "Kofi" are ARGUMENTS — actual values
greet("Amara")    # name receives "Amara"
greet("Kofi")     # name receives "Kofi"
greet("Ama")      # same function, different data each time
```

```python
# CODE EXAMPLE 10.2b: Multiple parameters

def calculate_total(price, quantity):
    total = price * quantity
    print(f"Price: {price} x Qty: {quantity} = Total: {total}")

calculate_total(25.50, 3)   # price=25.50, quantity=3
calculate_total(10.00, 5)   # price=10.00, quantity=5

# Order matters! These produce different results:
def introduce(first_name, last_name):
    print(f"My name is {first_name} {last_name}.")

introduce("Amara", "Mensah")   # correct order
introduce("Mensah", "Amara")   # reversed — still runs but wrong meaning
```

## 10.3 Return Values

Some functions **do** something (print output). Others **calculate** something and give back a result.

- `return` sends a value back to the **caller**
- When Python hits `return`, the function stops immediately
- If there is no `return`, the function implicitly returns `None`
- **Critical:** use `return` when you need the result for further use — `print()` only shows output, it does not pass data


```python
# CODE EXAMPLE 10.3a: return vs print

# With return — result can be used
def add(a, b):
    return a + b

total = add(10, 25)        # capture the returned value
print("Total:", total)     # 35
print(add(5, 3) * 2)       # 16 — used directly in expression

# Without return — returns None
def show_message():
    print("Hello!")

x = show_message()    # prints "Hello!" then x = None
print(x)              # None
# print(x * 2)        # TypeError — can't multiply None
```

```python
# CODE EXAMPLE 10.3b: return stops the function

def check_pass(score):
    if score >= 50:
        return "PASS"    # function ends here if score >= 50
    return "FAIL"        # only reaches here if score < 50

print(check_pass(75))    # PASS
print(check_pass(40))    # FAIL

# Returning multiple related values as a tuple
def min_max(numbers):
    return min(numbers), max(numbers)

low, high = min_max([5, 2, 8, 1, 9])
print(f"Min: {low}, Max: {high}")
```

## 10.4 Default Arguments

Default arguments make some parameters **optional**.

- Set a default with `=` in the function definition
- If the caller provides a value, that value is used — otherwise the default applies
- **Rule:** parameters with defaults must come **AFTER** parameters without defaults
- Violating this rule causes a `SyntaxError`


```python
# CODE EXAMPLE 10.4a: Default argument

def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Amara")                    # Hello, Amara! (uses default)
greet("Kofi", "Good morning")     # Good morning, Kofi! (overrides default)
greet("Ama", "Welcome back")      # Welcome back, Ama!

# Wrong order → SyntaxError (uncomment to see the error)
# def greet(greeting="Hello", name):   # SyntaxError!
#     print(f"{greeting}, {name}!")
```

```python
# CODE EXAMPLE 10.4b: Multiple default arguments

def create_profile(name, role="Student", active=True):
    status = "Active" if active else "Inactive"
    print(f"{name} | {role} | {status}")

create_profile("Amara")                          # uses both defaults
create_profile("Mr. Boateng", role="Teacher")    # override role only
create_profile("Kofi", "Student", False)         # override both
create_profile("Ama", active=False)              # keyword argument
```

## 10.5 Variable Scope — Local vs Global

**Scope** defines where a variable can be accessed.

- **Local variable** — defined inside a function; exists only while the function runs
- **Global variable** — defined outside all functions; readable from anywhere
- Functions can **read** global variables freely
- To **modify** a global variable inside a function, use the `global` keyword
- Best practice: avoid `global` — prefer passing values as parameters and returning results


```python
# CODE EXAMPLE 10.5a: Local vs global scope

counter = 0          # global variable

def increment():
    global counter   # declare intent to modify the global
    counter += 1

increment()
increment()
print(counter)       # 2

def local_example():
    local_var = "I am local"
    print(local_var)     # works fine inside the function

local_example()
# print(local_var)   # NameError! local_var does not exist out here
```

```python
# CODE EXAMPLE 10.5b: Better practice — parameters instead of global

# BAD — relies on global state
total = 0
def add_to_total(amount):
    global total
    total += amount

add_to_total(50)
add_to_total(30)
print("Total (global):", total)    # 80

# BETTER — pass in, return out
def add_amount(current_total, amount):
    return current_total + amount

balance = 0
balance = add_amount(balance, 50)
balance = add_amount(balance, 30)
print("Balance (clean):", balance)  # 80
# No global state — easier to test and reuse
```

## Topic 10 Recap

**Define:** `def function_name():` — body is the indented block beneath.

**Call:** `function_name()` — must be called explicitly to run.

**Parameters vs Arguments:** parameters are in the definition; arguments are in the call.

**Return:** `return value` sends data back to the caller; no return → `None`.

**Defaults:** `def f(x, y=10):` — default args must come after non-default args.

**Scope:** local variables exist only inside the function; use `global` sparingly to modify globals.

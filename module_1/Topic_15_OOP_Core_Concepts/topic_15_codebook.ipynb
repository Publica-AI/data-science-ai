## 15.1 Inheritance

- `class Child(Parent):` — Child inherits all attributes and methods from Parent
- Avoids repeating common code across multiple classes
- Child gets everything the parent has, plus its own additions
- Call `super().__init__(...)` in the child to run the parent constructor
- Terminology: Parent = base/superclass; Child = derived/subclass


```python
# CODE EXAMPLE 15.1a: Basic inheritance

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start(self):
        print(f'{self.make} {self.model} is starting...')

    def __str__(self):
        return f'{self.year} {self.make} {self.model}'

class Car(Vehicle):
    def __init__(self, make, model, year, doors=4):
        super().__init__(make, model, year)   # parent constructor
        self.doors = doors                    # child-specific attr

    def honk(self):
        print('Beep beep!')

class Truck(Vehicle):
    def __init__(self, make, model, year, payload_tons):
        super().__init__(make, model, year)
        self.payload_tons = payload_tons

    def load(self):
        print(f'Loading up to {self.payload_tons} tons.')

car = Car('Toyota', 'Camry', 2022)
car.start()     # inherited
car.honk()      # own method
print(car)      # inherited __str__

truck = Truck('Ford', 'F-150', 2021, 1.5)
truck.start()
truck.load()
```

```python
# CODE EXAMPLE 15.1b: Multi-level inheritance

class Animal:
    def __init__(self, name):
        self.name = name

    def breathe(self):
        print(f'{self.name} breathes.')

class Mammal(Animal):
    def feed_young(self):
        print(f'{self.name} feeds its young.')

class Dog(Mammal):
    def bark(self):
        print(f'{self.name} barks: Woof!')

# Dog inherits from Mammal which inherits from Animal
rex = Dog('Rex')
rex.breathe()      # from Animal
rex.feed_young()   # from Mammal
rex.bark()         # from Dog

print(isinstance(rex, Dog))     # True
print(isinstance(rex, Mammal))  # True
print(isinstance(rex, Animal))  # True
```

## 15.2 Method Overriding and super()

- **Method overriding:** redefine a parent method in a child class
- Python always uses the most specific version (child first)
- `super().method()` calls the parent's version from inside the child
- `super().__init__(...)` runs the parent constructor — call before adding child-specific attributes
- Overriding allows specialization while keeping the same interface


```python
# CODE EXAMPLE 15.2a: Method overriding

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f'Hi, I am {self.name}, age {self.age}.'

class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def introduce(self):          # overrides Person.introduce
        base = super().introduce()  # calls parent version
        return base + f' My ID is {self.student_id}.'

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def introduce(self):          # overrides differently
        return f'I am {self.name}, I teach {self.subject}.'

p = Person('Ama', 35)
s = Student('Kofi', 20, 'STU002')
t = Teacher('Mr. Boateng', 40, 'Python')

for person in [p, s, t]:
    print(person.introduce())
```

```python
# CODE EXAMPLE 15.2b: super() in __str__ override

class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def __str__(self):
        return f'{self.make} {self.model}'

class ElectricCar(Vehicle):
    def __init__(self, make, model, range_km):
        super().__init__(make, model)
        self.range_km = range_km

    def __str__(self):
        base = super().__str__()    # 'Tesla Model 3'
        return f'{base} (Electric, {self.range_km}km range)'

ev = ElectricCar('Tesla', 'Model 3', 570)
print(ev)    # Tesla Model 3 (Electric, 570km range)
```

## 15.3 Encapsulation

- **Encapsulation:** hide internal data to prevent accidental modification
- Prefix with `__` (double underscore) to make an attribute private (name-mangled)
- Access private data through **getter** methods and change via **setter** methods
- `@property` creates a getter that looks like attribute access
- `@attr.setter` creates a setter with validation
- Benefits: prevents invalid state, hides implementation details


```python
# CODE EXAMPLE 15.3a: Private attributes and getters/setters

class Student:
    def __init__(self, name, gpa):
        self.name = name
        self.__gpa = gpa         # private

    def get_gpa(self):
        return self.__gpa

    def set_gpa(self, value):
        if 0.0 <= value <= 4.0:
            self.__gpa = value
        else:
            raise ValueError(f'GPA must be 0.0–4.0, got {value}')

s = Student('Amara', 3.8)
print(s.get_gpa())   # 3.8
s.set_gpa(3.9)       # valid
print(s.get_gpa())   # 3.9

try:
    s.set_gpa(5.0)   # invalid
except ValueError as e:
    print(f'Error: {e}')

# Direct access fails (name-mangled):
# print(s.__gpa)   # AttributeError
```

```python
# CODE EXAMPLE 15.3b: @property decorator

class Circle:
    def __init__(self, radius):
        self.__radius = radius

    @property
    def radius(self):             # getter
        return self.__radius

    @radius.setter
    def radius(self, value):      # setter with validation
        if value <= 0:
            raise ValueError('Radius must be positive.')
        self.__radius = value

    @property
    def area(self):               # computed property (read-only)
        import math
        return math.pi * self.__radius ** 2

c = Circle(5)
print(c.radius)    # 5  — looks like attribute access
print(c.area)      # 78.54

c.radius = 10      # calls setter
print(c.area)      # 314.16

try:
    c.radius = -3  # raises ValueError
except ValueError as e:
    print(f'Error: {e}')
```

## 15.4 Polymorphism

- **Polymorphism:** 'many forms' — same method name behaves differently depending on the object type
- Python supports polymorphism naturally — no special syntax required
- Works through method overriding: each class defines its own version of a shared method name
- Enables writing generic code that works with any object that has the required method
- Duck typing: 'if it has the method, use it' — the class does not matter, only the interface


```python
# CODE EXAMPLE 15.4a: Polymorphism with shapes

import math

class Shape:
    def area(self):
        return 0
    def describe(self):
        return f'{self.__class__.__name__}: area = {self.area():.2f}'

class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return math.pi * self.r ** 2

class Rectangle(Shape):
    def __init__(self, w, h): self.w, self.h = w, h
    def area(self): return self.w * self.h

class Triangle(Shape):
    def __init__(self, b, h): self.b, self.h = b, h
    def area(self): return 0.5 * self.b * self.h

# Polymorphic — works with ANY Shape subclass
shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 8)]
for s in shapes:
    print(s.describe())   # each object uses its own area()
```

```python
# CODE EXAMPLE 15.4b: Duck typing — polymorphism beyond inheritance

# These classes are NOT related by inheritance
class Dog:
    def speak(self): return 'Woof!'

class Cat:
    def speak(self): return 'Meow!'

class Duck:
    def speak(self): return 'Quack!'

# Works with any object that has speak()
animals = [Dog(), Cat(), Duck()]
for a in animals:
    print(f'{a.__class__.__name__}: {a.speak()}')

# Python doesn't care about the class — only that speak() exists
# This is duck typing: 'if it quacks like a duck, it's a duck'
```

## 15.5 isinstance() and issubclass()

- `isinstance(obj, Class)` — True if obj is an instance of Class OR any subclass
- `issubclass(Child, Parent)` — True if Child inherits from Parent (directly or indirectly)
- Prefer `isinstance()` over `type(obj) == Class` — respects inheritance hierarchy
- Can check against multiple classes: `isinstance(obj, (ClassA, ClassB))`
- Used for safe type checking in functions that accept mixed object types


```python
# CODE EXAMPLE 15.5a: isinstance and issubclass

class Animal: pass
class Dog(Animal): pass
class GoldenRetriever(Dog): pass

d = Dog()
g = GoldenRetriever()

# isinstance — respects hierarchy
print(isinstance(d, Dog))              # True
print(isinstance(d, Animal))           # True
print(isinstance(g, Dog))              # True
print(isinstance(g, Animal))           # True

# type() — strict; doesn't respect hierarchy
print(type(d) == Dog)                  # True
print(type(d) == Animal)               # False — missing subclass

# issubclass
print(issubclass(Dog, Animal))         # True
print(issubclass(GoldenRetriever, Animal))  # True (indirect)
print(issubclass(Animal, Dog))         # False (reversed)
```

```python
# CODE EXAMPLE 15.5b: isinstance for safe processing

class Shape: pass
class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return 3.14159 * self.r ** 2
class Rectangle(Shape):
    def __init__(self, w, h): self.w, self.h = w, h
    def area(self): return self.w * self.h

def process(item):
    if isinstance(item, Shape):
        print(f'Shape area: {item.area():.2f}')
    elif isinstance(item, (int, float)):
        print(f'Number: {item}')
    elif isinstance(item, str):
        print(f'Text: {item}')
    else:
        print(f'Unknown type: {type(item)}')

items = [Circle(5), 42, 'hello', Rectangle(3, 4), [1, 2, 3]]
for item in items:
    process(item)
```

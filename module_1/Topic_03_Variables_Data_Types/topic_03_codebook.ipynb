## 3.1 What Variables Are

A **variable** is a named container that stores a value in memory.

Think of it as a labelled box:
- The **label** is the variable name
- The **contents** is the value
- The `=` sign is the **assignment operator** — it puts the value into the container

Key rules:
- `=` means **assign** (store a value)
- `==` means **compare** (check if two values are equal)
- Variable names should be descriptive — `student_age` is better than `x`


````python
# ============================================================
# CODE EXAMPLE 3.1a: Creating and using variables
# ============================================================

# Assign values to variables
student_name = "Kofi"       # text stored in a variable
student_age = 21            # number stored in a variable
gpa = 3.8                   # decimal number
is_graduated = False        # True/False value

# Use the variables to display information
print("Name:", student_name)
print("Age:", student_age)
print("GPA:", gpa)
print("Graduated:", is_graduated)
````

````python
# ============================================================
# CODE EXAMPLE 3.1b: Assignment (=) vs Comparison (==)
# ============================================================

score = 85           # = assigns the value 85 to 'score'

# == compares and returns True or False
print(score == 85)   # True  — score IS 85
print(score == 90)   # False — score is NOT 90
print(score == 85)   # True again

# Variables can be updated (reassigned)
score = 90           # now score holds 90
print("Updated score:", score)
````

## 3.2 Naming Conventions

Python follows **PEP 8** style guidelines. For variables, the rule is `snake_case`:
- All **lowercase** letters
- Words separated by **underscores** `_`

**Hard rules (will break your code if violated):**
- Cannot start with a number: `2name` is invalid
- Cannot contain spaces: `my name` is invalid
- Cannot use Python reserved keywords: `if`, `for`, `while`, `class`, `return`

**Best practices:**
- Use descriptive names: `student_total_score` beats `s`
- `camelCase` (like JavaScript) is discouraged in Python


````python
# ============================================================
# CODE EXAMPLE 3.2a: Correct and incorrect naming
# ============================================================

# --- CORRECT snake_case names ---
first_name = "Amara"
last_name = "Johnson"
student_id = 10245
is_enrolled = True
total_marks_available = 100

print(first_name, last_name)
print("ID:", student_id)
print("Enrolled:", is_enrolled)

# --- The names below would cause errors ---
# 2nd_name = "Johnson"   # SyntaxError: starts with a number
# my name = "Amara"      # SyntaxError: contains a space
# class = "Python"       # SyntaxError: 'class' is a keyword
````

## 3.3 Core Data Types: int, float, str, bool

Python has four fundamental data types you will use in almost every program:

**`int` (Integer)** — whole numbers, positive or negative, no decimal
- Examples: `42`, `-7`, `0`, `1000`

**`float` (Floating-point)** — numbers with a decimal point
- Examples: `3.14`, `-0.5`, `99.9`

**`str` (String)** — text, always enclosed in quotes (`""` or `''`)
- Examples: `"Hello"`, `'Python'`, `"42"` (number in quotes = string!)

**`bool` (Boolean)** — only two possible values: `True` or `False` (capital first letter!)
- Common mistake: writing `true` or `TRUE` — Python won't recognise these as booleans


````python
# ============================================================
# CODE EXAMPLE 3.3a: All four core data types
# ============================================================

# int — whole numbers
year_of_birth = 2002
number_of_courses = 5
temperature_low = -3

# float — decimal numbers
cgpa = 3.75
weight_kg = 68.5
pi = 3.14159

# str — text (must use quotes)
full_name = "Ama Owusu"
country = 'Ghana'
phone_number = "0244123456"   # digits in quotes = string, not a number!

# bool — True or False (capital first letter)
is_active = True
has_paid_fees = False

# Print them all
print(year_of_birth, type(year_of_birth))
print(cgpa, type(cgpa))
print(full_name, type(full_name))
print(is_active, type(is_active))
````

````python
# ============================================================
# CODE EXAMPLE 3.3b: The difference between '42' and 42
# ============================================================

number = 42          # integer — can do maths with it
text = "42"          # string — looks like a number but is text

print(type(number))  # <class 'int'>
print(type(text))    # <class 'str'>

# Maths works on int
print(number + 8)    # 50

# Adding a string 'concatenates' (joins text)
print(text + "8")    # "428" — not 50!
````

## 3.4 Type Checking & Casting

**Type checking** — use `type()` to find out what type a variable is:
```python
type(my_variable)   # returns the class/type
```

**Type casting** — converting a value from one type to another:
- `int(x)` — converts to integer (truncates decimals)
- `float(x)` — converts to floating-point number
- `str(x)` — converts to string

**Most important use case:** `input()` always returns a string. To do maths, cast it:
```python
age = int(input("Enter age: "))
```

**Danger:** You cannot cast a non-numeric string to a number:
```python
int("hello")  # ValueError!
```


````python
# ============================================================
# CODE EXAMPLE 3.4a: Using type() and casting
# ============================================================

# type() shows the data type
x = 3.99
print(type(x))              # <class 'float'>

# int() truncates (cuts off) the decimal
x_as_int = int(x)
print(x_as_int)             # 3 (not rounded, just truncated)

# str() converts anything to text
age = 25
message = "Age is: " + str(age)   # must convert int to str for + concatenation
print(message)

# float() converts int or numeric string to decimal
grade = "85"
grade_as_float = float(grade)
print(grade_as_float + 5.5)  # 90.5
````

````python
# ============================================================
# CODE EXAMPLE 3.4b: The standard input + cast pattern
# ============================================================

# Collect numeric inputs — convert immediately
name = input("Enter your name: ")
age = int(input("Enter your age: "))         # string → integer
height = float(input("Enter height in m: ")) # string → float

# Now we can do calculations
year_born = 2024 - age

print(f"\nHello, {name}!")
print(f"You were born around {year_born}.")
print(f"Height: {height} m")
````

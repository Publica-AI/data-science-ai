## 14.1 What is OOP

**Object-Oriented Programming** organizes code around objects.

- **Object:** a bundle of data (attributes) + behaviour (methods)
- **Class:** the blueprint that describes an object's structure
- **Instance:** a specific object created from a class
- OOP mirrors real-world thinking: a Dog has a name and can bark
- Benefits: organized, reusable, easier to extend and maintain


```python
# CODE EXAMPLE 14.1a: Procedural vs OOP comparison

# --- Procedural ---
# Data and functions completely separate
dog_name = 'Rex'
dog_breed = 'Labrador'

def dog_bark(name):
    print(f'{name} says: Woof!')

dog_bark(dog_name)

# --- OOP ---
# Data and behaviour bundled together in a class
class Dog:
    def __init__(self, name, breed):
        self.name = name      # data
        self.breed = breed    # data

    def bark(self):           # behaviour
        print(f'{self.name} says: Woof!')

rex = Dog('Rex', 'Labrador')  # create an instance
rex.bark()                    # call a method on the object
```

```python
# CODE EXAMPLE 14.1b: Multiple instances from one class

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def describe(self):
        print(f'{self.year} {self.make} {self.model}')

# Each instance is independent
car1 = Car('Toyota', 'Camry', 2022)
car2 = Car('Honda', 'Civic', 2021)
car3 = Car('Ford', 'Focus', 2020)

for car in [car1, car2, car3]:
    car.describe()
```

## 14.2 Defining Classes and Creating Objects

- `class ClassName:` — define a class (use PascalCase)
- `def __init__(self, ...):` — constructor; runs on object creation
- `self` — refers to the specific instance being created/used
- `object = ClassName(args)` — create an instance
- Each instance is independent — changing one does not affect others


```python
# CODE EXAMPLE 14.2a: Class definition and instantiation

class Student:
    def __init__(self, name, student_id, gpa):
        self.name = name
        self.student_id = student_id
        self.gpa = gpa

# Create instances
s1 = Student('Amara Mensah', 'STU001', 3.8)
s2 = Student('Kofi Boateng', 'STU002', 3.5)

# Access attributes with dot notation
print(s1.name)           # Amara Mensah
print(s2.student_id)     # STU002
print(s1.gpa)            # 3.8

# Modify an attribute
s1.gpa = 3.9
print(s1.gpa)            # 3.9 (only s1 changed)
print(s2.gpa)            # 3.5 (s2 unchanged)
```

```python
# CODE EXAMPLE 14.2b: isinstance() and type()

class Circle:
    def __init__(self, radius):
        self.radius = radius

c1 = Circle(5)
c2 = Circle(10)

print(type(c1))                    # <class '__main__.Circle'>
print(isinstance(c1, Circle))      # True
print(isinstance(c1, str))         # False

# Objects can be stored in data structures
circles = [Circle(1), Circle(5), Circle(10)]
for c in circles:
    print(f'Circle with radius {c.radius}')
```

## 14.3 Instance Attributes and Methods

- **Instance attributes:** variables set on `self` — unique to each object
- **Instance methods:** functions inside the class that use `self`
- Methods can read AND modify the object's own attributes
- Call methods with: `object.method(args)` — `self` is passed automatically
- Methods can call other methods using `self.other_method()`


```python
# CODE EXAMPLE 14.3a: Methods that read and modify attributes

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f'Deposited {amount}. New balance: {self.balance}')

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f'Withdrew {amount}. New balance: {self.balance}')
        else:
            print('Insufficient funds.')

    def get_balance(self):
        return self.balance

acc = BankAccount('Amara', 500)
acc.deposit(200)
acc.withdraw(100)
print('Balance:', acc.get_balance())
```

```python
# CODE EXAMPLE 14.3b: Methods calling other methods

import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def circumference(self):
        return 2 * math.pi * self.radius

    def scale(self, factor):
        self.radius *= factor

    def summary(self):
        # calls other methods using self
        return (f'Radius: {self.radius:.2f} | '
                f'Area: {self.area():.2f} | '
                f'Circumference: {self.circumference():.2f}')

c = Circle(5)
print(c.summary())
c.scale(2)
print(c.summary())
```

## 14.4 The __str__ Method

- `__str__` is a **special/dunder method** (double underscore on both sides)
- Python calls it automatically when you `print()` an object or call `str()` on it
- Without it, printing an object shows the memory address: `<__main__.Student object at 0x...>`
- Must `return` a string — do not use `print()` inside it
- Makes debugging and logging much easier


```python
# CODE EXAMPLE 14.4a: __str__ method

class Student:
    def __init__(self, name, gpa):
        self.name = name
        self.gpa = gpa

# Without __str__
s1 = Student('Amara', 3.8)
print(s1)    # <__main__.Student object at 0x...> — not helpful!

# Add __str__
class Student:
    def __init__(self, name, gpa):
        self.name = name
        self.gpa = gpa

    def __str__(self):
        return f'Student({self.name}, GPA: {self.gpa})'

s2 = Student('Amara', 3.8)
print(s2)        # Student(Amara, GPA: 3.8)
print(str(s2))   # same

# Works in lists too
students = [Student('Kofi', 3.5), Student('Ama', 3.9)]
for s in students:
    print(s)
```

```python
# CODE EXAMPLE 14.4b: __str__ in a complete class

class Product:
    def __init__(self, name, price, in_stock):
        self.name = name
        self.price = price
        self.in_stock = in_stock

    def apply_discount(self, percent):
        self.price *= (1 - percent / 100)

    def __str__(self):
        status = 'In Stock' if self.in_stock else 'Out of Stock'
        return f'{self.name} | GHS {self.price:.2f} | {status}'

p = Product('Python Book', 50.00, True)
print(p)                  # Python Book | GHS 50.00 | In Stock
p.apply_discount(20)
print(p)                  # Python Book | GHS 40.00 | In Stock
```

## 14.5 Class Attributes vs Instance Attributes

- **Instance attribute:** defined in `__init__` with `self.` — unique to each object
- **Class attribute:** defined in the class body outside any method — shared by ALL instances
- Access class attribute via `ClassName.attr` or `instance.attr`
- If you assign to `instance.attr`, Python creates a new instance attribute that *shadows* the class attribute for that object only
- Class attributes are useful for counters, constants, and shared defaults


```python
# CODE EXAMPLE 14.5a: Class vs instance attributes

class Student:
    school = 'Python Academy'   # CLASS attribute
    count = 0                   # class-level counter

    def __init__(self, name):
        self.name = name        # INSTANCE attribute
        Student.count += 1      # update class counter via class name

s1 = Student('Amara')
s2 = Student('Kofi')
s3 = Student('Ama')

print(s1.school)        # Python Academy (class attr)
print(Student.school)   # Python Academy (same)
print(Student.count)    # 3

# Shadowing: set instance attr that shadows class attr
s1.school = 'New School'
print(s1.school)        # New School (instance)
print(s2.school)        # Python Academy (class — unchanged)
```

```python
# CODE EXAMPLE 14.5b: Practical use of class attributes

class BankAccount:
    interest_rate = 0.05      # 5% — class attribute (same for all accounts)
    total_accounts = 0

    def __init__(self, owner, balance=0):
        self.owner = owner            # instance
        self.balance = balance        # instance
        BankAccount.total_accounts += 1

    def apply_interest(self):
        interest = self.balance * BankAccount.interest_rate
        self.balance += interest
        return interest

    def __str__(self):
        return f'{self.owner}: GHS {self.balance:.2f}'

a1 = BankAccount('Amara', 1000)
a2 = BankAccount('Kofi', 2000)

a1.apply_interest()
a2.apply_interest()
print(a1)    # Amara: GHS 1050.00
print(a2)    # Kofi: GHS 2100.00
print('Total accounts:', BankAccount.total_accounts)   # 2
```

## Topic 14 Recap

**Class:** `class Name:` — blueprint for objects (PascalCase).

**Constructor:** `def __init__(self, ...):` — runs on object creation; sets instance attributes.

**Instance attributes:** `self.name = value` — unique to each object.

**Methods:** `def method(self):` — use `self` to access/modify the object's data.

**__str__:** `def __str__(self): return f'...'` — human-readable representation.

**Class attributes:** defined in class body (not in __init__) — shared by all instances.
